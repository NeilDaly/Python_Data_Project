# Tuples
___

- EG: `('python', 'sql', 'excel')`
- Now time to learn about the last container Data Type called Tuples
- We need to be aware of this one
- This one will have a lot of Similarities with Lists - As Tuples are a Sequenced Data Type
- However, there's one big difference
- Tuples are defined by Parenthesis `()` and then you Add the Items in there. Again, it can be any Item but we will predominently use Strings
- Unlike Lists however if we wanted to Modify a Tuples, we are not able to do things to it 
- We are unable to Remove, Add or even Append to. This is because Tuples are Immutable

| Feature          | List                              | Dictionary                             | Set            | Tuple                                    |
|------------------|-----------------------------------|----------------------------------------|----------------|------------------------------------------|
| Usage            | Collection of ordered items       | Key-value pairs                        | Unique items   | Fixed data                               |
| Syntax           | `[item1, item2, ...]`             | `{'key1': value1, 'key2': value2}`     | `{item1, ...}` | `(item1, item2, ...)` or `item,`         |
| Type of Data     | Sequence                          | Mapping                                | Set            | Sequence                                 |
| Order            | Ordered                           | Unordered                              | Unordered      | Ordered                                  |
| Indexing         | Yes (by index)                    | Yes (by key)                           | No             | Yes (by index)                           |
| Duplicate Values | Allowed                           | Values allowed, keys must be unique    | Not allowed    | Allowed                                  |
| Mutability       | Mutable                           | Mutable                                | Mutable        | Immutable                                |

- Tuples are for **Fixed Data**
- So if we needed to have a unique Set of Items that we don't want to be able to alter, we'd put them inside a **Tuple**
- What is the applicability of this?
- It's more applicable in Software Engineering where say we have a Unique Set of Items that we don't want to change
- They use that a lot in Software Engineering, in Data Analytics we will see this less  but it does come up from time to time

```python
# Let's say we have a Unique set of Skills and we aren't planing on learning any New Skills any time soon
# We can Set this Variable equal to a Tuple - Defined by the Parentheses 

neils_skills = ('python', 'sql', 'statistics', 'tableau')
neils_skills
```

```python
# Tuples are Indexed as well so we are able to Call things by their Indexes

neils_skills[0]
```

```python
# Similarly we can Access things by Slicing
# Here we are Acessing the first 2 Items in the Tuple
# We can use the same Slicing Method that we used in Lists

neils_skills[0:2]
```

```python
# And to prove that these Tuples are Immutable we can try and Invoke the Append Method as seen below
# This will result in an Error - Tuple Object does not have the Attribute of Append

neils_skills.append('r')
```

```python
# So, what Attributes does Tuple have?
# We can use the good oul reliable Help Function to find out!!

help(tuple)
```

```python
# Tuples are Immutable
# But, let's say we do need to actually Update the List of Items in here 
# How can we do this with Operators?
# Let's dive into it below
# So we have our List of Skills in our Original Tuple and we now have a New List of Skills as another Tuple

neils_skills = ('python', 'sql', 'statistics', 'tableau')
print(id(neils_skills))

new_skills = ('excel', 'power bi')

# We could technically just Add the 2 Tuples together
# Running this, we get this New Tuple List back 
# Even though it's immutable, we aren't changing it
# We are Creating a whole New Object whenever we modify the Tuple. Think of this as a Loop hole. It's a way of getting around the Rules of Tuples
# Comparing the 2 Id's of the Objects, we can see they are different which proves, these are completely different Objects even though they have the same Names

neils_skills += new_skills
neils_skills
id(neils_skills)
```

```python
# Doing the exact same here as we did above
# Except we are using Lists so we can see that the ID's are the same and that they are the same Objects

neils_skills = ['python', 'sql', 'statistics', 'tableau']
print(id(neils_skills))

new_skills = ['excel', 'power bi']

# Comparing the 2 Id's of the Objects, we can see they are the same, which proves these are identical Objects

neils_skills += new_skills
neils_skills
id(neils_skills)
```

```python
# Mutliple ways we can do it. I'm just adding them down here so we can see
# They all do the same things, but it's good to see the various ways to do this

# Just having this will Add them together and Print the New Tuple out
neils_skills + new_skills

# Having this, will Add them Together but Assign the New Tuple to the Variable Name
# Then we Print out the New Tuple 
skills = neils_skills + new_skills
skills

# We can also do it with the Plus and Assign Operator as seen below
# Then we Print it out as well by Calling it

neils_skills += new_skills
neils_skills
```

```python
# In an Upcoming Section we will be looking at the Range() Container Data Type
# We will be looking at this one in the Loops section
# What does the range() Function do?
# We put a Value inside of there and that is the stop position

# So if we do the below, it'll Return the Range of (0, 5)
range(5)

# But what exactly is going on with the Range Function? To see this, we will have to Convert it into a Tuple to see what's going on inside of here
tuple(range(5))

# Running this, we can see that it's actually Values 0, 1, 2, 3, 4
# It doesn't do 5 as 5 is exclusive for the stop - In this case

# We can also do things by providing it the start Value. So in this case we want it to start at 2 and running it again we get 2, 3, 4
tuple(range(2,5))

# This is really convenient if we need to iterate through a certain set of numbers
# We want to go from 1 to 100 and in this case we only want the odd numbers
# Running the below Returns a Range of all the different Odd Numbers from 0 to 100
# The last Argument we Added into the below is the Step Value - What we can increment this start and stop by

tuple(range(1, 101, 2))
```
