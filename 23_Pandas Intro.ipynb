### Brief Introduction into Data Frames:
- Whenever we Create the above `Object`, (`pd.read_csv('...')` this Data Frame we need to name it into a Variable)
    - It's common when we are just doing one Data Frame, to name it `df`
- Data Frames are: Two-dimentional, size-mutable, potentially hertogeneous tabular data
    - Data Structure also contains labeled axes: IE: Rows and Columns
    - Arithmetic Operations align on both Row and Column Labels
    - Can be thought of as a dict-like container for Series Objects
    - This is the Primary Data Structure

```python
# As we can see below, the Data is laid out in Rows and Columns (like Excel)
# It also has an Index (Far left Column)
df
```

```python
# Let's start with the Columns first!
# Because it's a dict like Object, we can use a similar notation in order to access the Columns (lets access total_rooms)
# This Returns all the different Values inside that Column (Column Header is the Key:Column Values are the Values)

df['total_rooms']
```

```python
# Above is not the only way we can Access this Column Data, we can use dot notation (df.) to access it too
# We can explicitly call out the column name itself to get the Values as well (Will provide exactly the same Results)
df.total_rooms
```

- So we talked about accessing the Columns... but what about if we want to access the Index??
- Let's say we want to access the first item in `total_rooms`
    - Similar to that List Index, we can provide square brackets `[]` and then type in the Index Number to get the corresponding Value that we want to see from the named Column

```python
df.total_rooms[0]
```

```python
# This is the other way to do it but we get the exact same Results. Doesn't matter which way we do it
df['total_rooms'][2999]
```

```python
# And finally, let's download the Data
dataset = load_dataset('lukebarousse/data_jobs')
```

```python
# Now let's access it and because it's a Dictionary like Object we need that notation along with the Key Name which in this case is train
dataset['train']
```

```python
# Running the Type Function on this we get back a Dataset... we want it as a Data Frame, since that's how we will Load it into Pandas
type(dataset['train'])
```

```python
# To do this, we are going to Run a Method on this, that dataset actually has!
# We can transform it into a multitude of different things (csv, dict, etc) but we are going to Transform it into Pandas!
# And now, we have it Imported in as a Dataframe!!
df = dataset['train'].to_pandas()
df
```
