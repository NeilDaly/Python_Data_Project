# Conditional Statements
___

- Conditional Statements are `IF` Statements
- In here, we are going to look at: `IF`, `ELIF`, `ELSE`
- These Statements do very simple things, such as Evaluating weather if a Condition is True or False and then executing it based on that

| Part   | Purpose                           | Condition Evaluated              | Action if True                                   |
|--------|-----------------------------------|----------------------------------|--------------------------------------------------|
| `if`   | First Condition Check             | First Condition to be Evaluated  | Executes Block if Condition is True              |
| `elif` | Subsequent Condition Check        | Next Condition, if `if` is False | Executes Block if this Condition is True         |
| `else` | Fallback if no Conditions are met | None                             | Executes Block if all above Conditions are False |


```python
# Start a Conditional Statement by defining a Keyword
# It will always start with the IF Keyword
# Then the Condition follows
# And then the Colon, which defines that it's a Function to be executed
# Now for the action that we want to Complete
if True:
    print("What's up Data Nerds?")
```

```python
# Start a Conditional Statement by defining a Keyword
# It will always start with the IF Keyword
# Then the Condition follows
# And then the Colon, which defines that it's a Function to be executed
# Now for the action that we want to Complete
if False:
    print("What's up Data Nerds?")
```

```python
# Now for a more practical Example of how you'd see this Statement Operated in the real world:
    # We are trying to Filter out Applicants for a Job based on a Skill
    # Specifically, we are looking for Applicants with the Job Skill of SQL

# Applicant Skill
applicant_skill = "SQL"

# Skill Required for Job Posting
job_skill = "SQL"
```

```python
if applicant_skill == job_skill:
    print("We have found an Applicant that matches the Job")
```

```python
# Now we are going to look at an elif Example
# Now we want to check years of Expierience
# Now we have a Condition where we are checking the Skill of SQL
# But let's say they don't match this Condition
# But say they have a lot of years of Expierience and so they would still be something that we wanted to match and flag

# Applicant Skill
applicant_skill = "Python"

# Skill Required for Job Posting
job_skill = "SQL"

# Number of years Expierience
years_expierience = 6

if applicant_skill == job_skill:
    print("We have found an Applicant that matches the Job")
elif years_expierience >= 5:
    print("Has the Expierience but lacks the Skills")

```

```python
# What happens if we have a Condition that is accounted for?
# What happens if we put a 3 in years_expierience and Run the Code - Nothing happens because we have nothing to handle this scenario
# Because none of the COnditions will be True, it ends up fizziling out because we didn't account for this Scenario
# Enter the Else Keyword
# Now, when all the above conditions Fail, the Else will be invoked

# Applicant Skill
applicant_skill = "Python"

# Skill Required for Job Posting
job_skill = "SQL"

# Number of years Expierience
years_expierience = 3

if applicant_skill == job_skill:
    print("We have found an Applicant that matches the Job")
elif years_expierience >= 5:
    print("Has the Expierience but lacks the Skills")
elif applicant_skill == "Python":
    print("This applicant should know SQL...")
else:
    print("Not a good candidate")
```

```python
# Sometimes you may see the keyword Pass inside people's Functions
# This is typically put into place whenever you don't have a value you want to put there or don't have one
# You can use this Pass to fill in for a Value so your Code Runs and doesn't throw Syntax Errors
# Since you will need something in there for the Function to Run
# Below, we showcase this, so when we Run this, nothing will happen because Pass is in there but it still executes

# Applicant Skill
applicant_skill = "Python"

# Skill Required for Job Posting
job_skill = "SQL"

# Number of years Expierience
years_expierience = 3

if applicant_skill == job_skill:
    print("We have found an Applicant that matches the Job")
elif years_expierience >= 5:
    print("Has the Expierience but lacks the Skills")
elif applicant_skill == "Python":
    pass
else:
    print("Not a good candidate")
```
