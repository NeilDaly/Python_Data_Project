```python
import pandas as pd
from datasets import load_dataset

# Load the Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
# Only way to include DateTime information as just Running the .describe() won't bring it up since that way is Depreciated
df.describe(datetime_is_numeric=True)
```

```python
df.info()
```

- Anytime we are starting to do any Analysis, the 2 Methods that should be used first are:
    - `df.describe()`
    - `df.info()`

- There are a lot of different Methods that can be used to Analyze Data on the Data Frame and even the Columns themselves
    - This is to better understand the Key Statistics about a Column
    - **Note: We aren't going to do all of these, we are only going to do a few**
        - `df.count()` - Count of Values
        - `df.sum()` - Sum of Values
        - `df.cumsum()` - Cumulative Sum of Values
        - `df.min()/df.max()` - Minimum/Maximum Values
        - `df.idxmin()/df.idxmax()` - Indexes of MIN/MAX Values
        - `df.mean()` - Mean of Values
        - `df.median()` - Median of Values
        - `df.mode()` - Mode of the Values
        - `df.column_name.value_counts()` - Unique Counts

```python
# As a point, the info() Method already shows a Count of all of those Columns but we can also Run:
df.count()
# Which gives us the same Number of Results
```

```python
# Moving to more of a Numerical Analysis, we want to get the Median Value, we Run:
df.median()
# This Value is not included in the describe() Method
```

- Regarding the `describe()` Method, when we Ran it before, it provided a lot of the Key Statistics in there
    - But specifically with the Min and Max, sometimes when this is provided, we would want to go in further and investigate what's going on here because we may have very strange Values
    - In this case, we have a Minimum Salary of $15,000 which is extreamly low... so, let's investigate!

```python
# First to Confirm that Value, we are going to Run the Min Method on this
df['salary_year_avg'].min()
```

```python
# Now to Run the other Method of idxmin(). This gets the Index Value of where this Minimum Salary Lives
# It's essentially a CTRL + F
min_salary = df['salary_year_avg'].idxmin()
```

```python
# Now to access this and get back the Record, use the iloc Property on this Data Frame and specify the Index of the Minimum Salary (which is inside our Variable)
df.iloc[min_salary]
```

    - Our findings indicate:
        - No, a $15,000 USD monthly salary is not "normal" in Brazil—it is extremely high and places an individual well within the top 1% or even top 0.1% of earners. 
        - With that income, you would be living a luxurious, upper-class lifestyle, likely far exceeding the salary of senior executives or specialized professionals in Brazil.
- Because of these findings, this isn't an anomoly!

```python
# To get back an Array of all the unique job titles, simply use:
df['job_title_short'].unique()
```

- While this is cool and all, it doesn't really provide much Value...
- We want to find out, what are the Unique Value Counts associated with these Jobs!
    - And for this, the `.value_counts()` Method will do the trick!

```python
df['job_title_short'].value_counts()
```

- Not only does it provide, Numerical Counts but it also Organizes it in DESC Order!
    - So we can see that **Data Analyst**, **Data Engineer**, and **Data Scientist** take up the majority of all the Roles and are the 3 most highest
    - Then we have **Business Analyst**, and **Software Engineer**, followed by the Senior Roles of **Senior Data Engineer**, **Senior Data Scientist**, and **Senior Data Analyst**
    - Then finally, we close this out with **Machine Learning Engineer** and **Cloud Engineer**

- We have beein using these Statistical Methods, but we've only really used it on one off Calculations...
- What happens, if we want to dive deeper now into understanding multiple different aspects...
    - Like, what is the miniumum salary of all of the different job_title_shorts?
- Well, as you guess it! There is a Method for this!
    - And it's `pandas.DataFrame.groupby`. The Syntax is as follows:
    - `df.groupby('columns to group')['columns to aggregate'].your_agg_method()`
- After we have Grouped our Data, we can perform a lot of the different Methods, that we mentioned at the start of this Lesson such as Min, Max, Count...

```python
# Question: What is the Miniumum Salary of all of the different job_title_shorts?
df.groupby('job_title_short')['salary_year_avg'].min()
```

- We not only see that lowest Role that we found earlier: The **Data Engineer** at $15,000, but also a whole host of other ones in there!
- Now that we have that, let's answer another Question that is a better representation of the Data itself.

```python
# Question: What is the Median Salary of all of the different job_title_shorts?
df.groupby('job_title_short')['salary_year_avg'].median()
```

- Let's manipulate this even further...
- We learned that there are Fluctuations in Salaries to different Countries so we can also do a Group By, not only by the **job_title_short** but also by the **Country**
    - To do this, the Syntax is as follows:
    - `df.groupby(['list of columns to group'])[['list of columns to aggregate']].your_agg_method()`

```python
# Question: What is the Median Salary of all of the different job_title_shorts in all the different job_country?
df.groupby(['job_title_short', 'job_country'])['salary_year_avg'].median()
```

- Now that we have this, we learned how to do multiple Group By's of the different Columns that we want to do
    - What if we want to perform Aggregations on multiple columns?
    - Because we have access to **salary_year_avg** and the **salary_hour_avg**, what happens if we want to look at both?

```python
# Question: What is the Yearly Median Salary and Hourly Median Salary of all of the different job_title_shorts?
df.groupby('job_title_short')[['salary_year_avg', 'salary_hour_avg']].median()
```

- Last thing to cover with the Group By:
    - Yeah we can get the Minimum Values of the Job Titles for the Yearly Salary, but what happens if we also want the Max Values in there too??
    - So we want to do multiple different operations or Aggregation Methods on the Data...
        - Well, ofc! There's a Method for that!
        - It's called `.agg()` and inside of there we need to provide a List of Arguments
    - To do this, the Syntax is as follows:
    - `df.groupby(['list of columns to group']).agg(['list of aggregation methods'])`

```python
# Question: What is the Yearly Min, Max, Median Salary of all of the different job_title_shorts?
df.groupby('job_title_short')['salary_year_avg'].agg(['min', 'max', 'median'])
```
