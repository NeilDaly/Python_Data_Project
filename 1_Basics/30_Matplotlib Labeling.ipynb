```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Load Dataset
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
# Getting the title_count series
title_count = df.job_title_short.value_counts()

# Plotting the Graph
plt.bar(title_count.index, title_count)
plt.show()
```

- Let's start simple. 
    - Let's **Add a Title**
    - There is a Function for that!
        - `matplotlib.pyplot.title`: All it does is, Sets a Title
        - The only thing that needs to be provided is a Label or a String
- It's important to Note, for Titles, Legends and even the Show Function, it's not done in an OOP Approach (Object Orientated Programming Approach)
    - MatplotLib enables what is called a Statefull Approach
        - This means, we can stack up all the different commands that we want to provide for it
        - In this case we want a Title so we can just Call it!
    This Statefull Approach is common in MatLab and is a common Approach in there and it's how it's going to be done in here!
- Now that we know that, let's Add some more Labelling and Cleanup/Documentation to this Visualization
    1. Give this Visualization a Title
    2. Give a Label to the Y Axis
    3. Fix the X Axis and make it Readable (Labels to 45 Degrees)

```python
# Plotting the Graph
plt.bar(title_count.index, title_count)

# Give it a Title
plt.title('Potings by Job Title')

# Y Axis Label
plt.ylabel('Count of Job Postings')

# Rotate the X Axis Labels 45 Degrees and Horizontally Align them (ha) to the Right of the Tick Mark
plt.xticks(rotation=45, ha='right')
plt.show()
```
