# Lists
___

- EG: `job_skills=['SQL', 'excel', 'python']`
- Lists are denoted by their square brackets `[]`, and can contain any Data Type inside them
- In this section we will be going over the common **Methods** associated with them such as Altering them, Adding to them and changing them
- We will also dive into unpacking Lists so we can Access individual Items inside of them

| Feature          | List                              | Dictionary                             | Set            | Tuple                                    |
|------------------|-----------------------------------|----------------------------------------|----------------|------------------------------------------|
| Usage            | Collection of ordered items       | Key-value pairs                        | Unique items   | Fixed data                               |
| Syntax           | `[item1, item2, ...]`             | `{'key1': value1, 'key2': value2}`     | `{item1, ...}` | `(item1, item2, ...)` or `item,`         |
| Type of Data     | Sequence                          | Mapping                                | Set            | Sequence                                 |
| Order            | Ordered                           | Unordered                              | Unordered      | Ordered                                  |
| Indexing         | Yes (by index)                    | Yes (by key)                           | No             | Yes (by index)                           |
| Duplicate Values | Allowed                           | Values allowed, keys must be unique    | Not allowed    | Allowed                                  |
| Mutability       | Mutable                           | Mutable                                | Mutable        | Immutable                                |

```python
# This is how we define a List
# We can put anything inside of a List, even another List!

my_list = [1, 'python', [2, 'SQL']]

my_list
```

```python
# Here is our practical example

job_skills = ['SQL', 'tableau', 'excel']
```

```python
# Another handy Method to use is the Length Method
# Inside of here, you can put any type of Object
# You can also add in a String to see the Character Count of it

len(job_skills)
```

### Indexes of a List

- Speaking of getting a Result of 3, let's talk about the Indexes of a List
- So for the List we are using, we have 3 Items
- They are Indexed as 0, 1, 2
- In Programming, we don't start at 1, we start at 0
- So let's look at why we do this and put the Indexes to use!

```python
# So for this List of Skills, let's get the second item in this List
# We can do this by using the Square Brackets and looking up the Index

job_skills[1]
```

- We need to Understand how Indexes work and what they are to use the Insert Method
- This is because the Insert Method works by Inserting an Object **BEFORE** an Index
- So if we want to put something in a certain place, we will need to know the Index of where it is going and what is before/after it
- It's Importnat to Remeber that Lists are **Ordered**

```python
# So for these Job Skills, let's now put Tableau back in
# We provide the Index of the Item that we want to Insert BEFORE
# Because we want it in the second position (where we had it before), we put it before Index 1
# Because Index 1 is the Second Item in the List
# It's easier to think of put the Index in of where you want it to go. Because we wanted it in the second position, we put in the Index of 1

job_skills.insert(1, 'tableau')
job_skills
```

```python
# Now, let's Remove Python from the List using a New Method that requires us knowing the Index
# The power of this is, the remove Method removes the first occurancce of the String
# The Pop Method, removes the Item at a certain Index

job_skills.pop(3)
```

```python
# Now Printing out our List, we can see the Last Item(pyhon) has been Removed
job_skills
```

### Slicing

- **Syntax:** `list[start:end:step]`
- `start`: The Starting Index (inclusive)
- `end`: The Ending Index (exclusive)
- `step`: Steps to take between Items

- The Index Values don't just have the Ability to access them 1 by 1
- Maybe we want to Access multiple Values of them
- Accessing multiple Values is done by ways of something called **Slicing**

```python
job_skills = ['sql', 'excel', 'python']
```

```python
# Let's just begin with the Start and End Values
# We want to get all Items in the List we do the below

job_skills[0:3]
```

```python
# Now, both of those Values are actually the Default Values of Start/Stop
# So, technically, if we want to get the entire List back all we need is the below

job_skills[:]
```

```python
# This is how we get only the first Item in the List

job_skills[0:1]
```

```python
# But because the first is a Default Value, we can leave it blank in this case

job_skills[:1]
```

```python
# Now let's go over the Step
# Here, it Defaults to 1
# So Running it with a Blank Start/End because the Default Values are the first and Last Values, running the below, we get all the Items in the list

job_skills[::1]
```

```python
# Now changing that Value to 2, to get every second one or every other Value we get

job_skills[::2]
```

```python
# Now changing that Value to 3, to get every third one 
# We only get back the First Item in this Example because there aren't enough Items in the List to actually do that

job_skills[::3]
```

```python
# Now let's do this on a bigger List so we can show the effects of this

# Define the 2 Lists
luke_skills = ['python', 'bigquery', 'r']
kelly_skills = ['python', 'sql', 'looker']

# Add the 2 Lists together to make 1 big List
all_skills = luke_skills + kelly_skills

all_skills

# One thing to Note, the Big List has Repeating Values in it
# That's okay because Duplicates are Allowed with Lists
# Also, it maintains the Order at which we Added those 2 Lists together since Lists are Ordered
```

```python
# Now we want to see every Second Item in that List

all_skills[::2]
```

```python
# What happens if we only want the Last Item in the List?
# What we can do is have the Start Value = to the Last Index in the List, in this Case, it's 5. Now we Start looking at Index 5

all_skills[5:]
```

```python
# Well conviniently the Indexes of these Lists are in Reverse but for their Negative Values
# We don't have to start at 0 because 0 is denoted for the first Item in the List so we start at -1
# Think of it as the first last Item
# In our example, the first last Item is Looked so that would be at Index -1 as seen below
# This is a much easier way to get the last Item in the List

all_skills[-1:]
```

```python
# And ofc, if we want to get the last 2 items, we can put in -2 and so on...

all_skills[-2:]
```

### Unpacking
___
Assign each Value in an iterable to a Variable in a single statement

- The last concept to cover is **Unpacking**
- This Involves assigning each of the Variables an Iterable

```python
job_skills = ['python', 'excel', 'sql', 'looker']
```

```python
# If we wanted to Assign each one of those Values from the List to it's own Variable

skill1, skill2, skill3 = job_skills
```

```python
# Calling a Print Statement on each one of those, we can see we have each of the Skills from the List, Assigned to a Variable
# They are all inside their own individual Variables and we did this all in 1 Line - This is called Unpacking

print(skill1)
print(skill2)
print(skill3)
```

```python
# Now, let's say we only care about that first Item in the List and we didn't really care about all the rest of them
# What we can do here, is use an Unpack Operator which is just an Asterisks (*) also called a STAR OPERATOR
# What's going to happen here is the Asterisk unpacks an iterable into the Variable of skill_we_dont_care_about
# In this case, skill_we_care_about will put the first Item into here and then all the rest of the Skills, will go into skill_we_dont_care_about
# Because we have the Unpack Operator in front of it
# This turns, the Second Variable into a List

skill_we_care_about, *skill_we_dont_care_about = job_skills
```

```python
# Now Printing it out to seeeeee

print(skill_we_care_about)
print(skill_we_dont_care_about)
```

- It's important to Note, this Unpack Operator is an Advanced Concept but it does come up from time to time
- Especially when you need to dive in to understand Functions which we will get to more in an upcoming Lesson so we need to be aware of it
- But we don't need to be a Master of it
