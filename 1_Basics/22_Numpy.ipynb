```python
# Create a List of 1,000,000 Salaries ranging from 50,000 to 150,000
import random

# Now for some List Comprehension to make a List of all these different Random Values
salary_list = [random.randint(50_000, 150_000) for iterable_Value in range(10_000_000)]
```

```python
# Now with this Salary List, say we want to find something of Analytical Importance such as the Mean or Average
# We can use the Statistics Libraby in order to figure that out

import statistics

# Running this we get the Mean of the List
statistics.mean(salary_list)

# But we don't really care about the answer... we only care about how long it's taking so we can compare Runtimes to Numpy
# Here's how we do that: Magic Command of timeit - This will Run the Statement Multiple times and provide us the Average Time it takes for Execution
# Placing the comments in here because Magic Commands need to be the first line to Execute
```

```python
%%timeit
statistics.mean(salary_list)

# This took 3.7 Seconds to Run whereas Numpy took .5 of a Second to Run - That's a massive difference!
```

```python
# Now let's do this with Numpy
import numpy as np

# Now we also want to time Numpy to find the Mean with the timeit Command
```

```python
%%timeit
np.mean(salary_list)
```

- This Runtime is a massive difference. Bigger the Data the less time you'll want to wait

```python
# Now we are going to look at some basic Operations using Numpy
# Define a core data types inside of Numpy in order to do Operations and that is - Array

# Created an Array and passed in the Object of a List of Numbers
my_array = np.array([1, 2, 3, 4, 5])
```

```python
# Get the MEAN of the Array
my_array.mean()
```

```python
# Array is the Basic Data Type of Numpy and we can do a lot of things with it
# Now let's move to a more practical example: We are working with Data Sciemce job Postings and we have our Data inside of Arrays

# Job Titles
job_titles = np.array(['Data Analyest', 'Data Scientist', 'Data Engineer', 'Machine Learning Engineer', 'AI Engineer'])

#Base Salaries
base_salaries = np.array([60_000, 80_000, 75_000, 90_000, None])

#Bonus Rates
bonus_rates = np.array([.05, .1, .08, .12, None])
```

```python
# What's cool about Arrays is, we can do computations between Arrays
# We want to find out what the Total Salary is based on the Base Salaries and Bonus Rates but we wanted to get the Array from that

total_salaries = base_salaries * (1 + bonus_rates)

total_salaries
```

```python
# Now if we want to see the Mean of these Total Salaries

np.mean(total_salaries)
```

- You can imagine, if we had hundreds of thousands of different Values inside something like Numpy Arrays, we can do our Calculations a lot quicker then if they were inside a Standard List inside Python
- Let's say we have some Values insode of our Data that we have missing Data or NULL Values, as in we collect a Job Title, but we don't know the Salary
- We Add in a fifth job of `AI Engineer` but we won't put in a Salary for it
- Python has this Type of `None`. If we add it in, it won't Print anything out as it's nothing - We will Add this into the Array as the 'Blank' Value
- Problem is... when we add it in and try and Execute the `total_salaries` Method, we get an Error... because we can't Add 1 + NoneType to each other
- So what would be a better way to store that we have No Values??
    - Well Numpy has a Function called `nan` - This means Not A Number. This is specifically for Numerical Data
    - Because it's recognized as a `Float`, we can Add `INT`'s and `Float` together and not get an Error
    - We Replace that `None` with `nan`
    - We also replace the Function as `nanmean` as this is built to handle nan
    - Now after Running it, our Mean is the same, it did not change, which means the `nan` Value was Ignored and didn't bring the Mean down and didn't influence it in any way

```python
# Array is the Basic Data Type of Numpy and we can do a lot of things with it
# Now let's move to a more practical example: We are working with Data Sciemce job Postings and we have our Data inside of Arrays

# Job Titles
job_titles = np.array(['Data Analyest', 'Data Scientist', 'Data Engineer', 'Machine Learning Engineer', 'AI Engineer'])

#Base Salaries
base_salaries = np.array([60_000, 80_000, 75_000, 90_000, np.nan])

#Bonus Rates
bonus_rates = np.array([.05, .1, .08, .12, np.nan])
```

```python
total_salaries = base_salaries * (1 + bonus_rates)

total_salaries
```

```python
np.nanmean(total_salaries)
```
