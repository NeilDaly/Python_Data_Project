### Logical Operators

- EG: `and, or, is, is not`
- For this section we are going to be covering Operators - We will be covering Logical Operators first: `and, or, not`
- What is the purpose of the Operators? They can be Applied in multiple different ways

| Operator | Name        | Example |
|----------|-------------|---------|
| and      | Logical AND | x and y |
| or       | Logical OR  | x or y  |
| not      | Logical NOT | not x   |

```python
# AND Example
# Created 2 Variables and set them both to TRUE
job_work_from_home = True
job_health_insurance = True

# In this case we are Comparing weather both Values are TRUE and if they are, Return TRUE
job_work_from_home and job_health_insurance
```

```python
# Created 2 Variables and one to TRUE and one to FALSE
job_work_from_home = True
job_health_insurance = False

# In this case we are Comparing weather both Values are TRUE and if they are, Return TRUE, if not Return FALSE
job_work_from_home and job_health_insurance
```

```python
# We aren't Limited to comparing 2 Values, we can Compare as many as we want
# Created 3 Variables and set them both to TRUE
job_work_from_home = True
job_health_insurance = True
job_no_degree_mention = True

# In this case we are Comparing weather ALL Values are TRUE and if they are, Return TRUE, if not Return FALSE
job_work_from_home and job_health_insurance and job_no_degree_mention
```

```python
# OR Example
# Created the below Variables
job_work_from_home = False
job_health_insurance = False
job_no_degree_mention = True

# In this case we are Comparing weather one of the Values are TRUE and if they are, Return TRUE (doesn't matter which Value is True)
# Only if ALL Values are FALSE, will we get FALSE Returned
job_work_from_home or job_health_insurance or job_no_degree_mention
```

```python
# NOT Example
# NOT just goes through and changes which way where the TRUE/FALSE Statement is going
# In this case we are putting NOT TRUE so it will Return FALSE

not True
```

```python
# NOT Example
# NOT just goes through and changes which way where the TRUE/FALSE Statement is going
# In this case we are putting NOT FALSE so it will Return TRUE

not False
```

```python
# Created the below Variables
job_work_from_home = False
job_health_insurance = False
job_no_degree_mention = True

# In this case we care about the opposite of weather a Degree is mentioned
# We put a NOT Statment in there so it will Return FALSE
# Checking weather either of WFH or Insurance are True (They aren't so that is False), then checks if Degree Mentioned is FALSE (it isn't, so then it Finally Returns FALSE)
job_work_from_home or job_health_insurance or not job_no_degree_mention
```

```python
# These Operator Keywords will be used in a lot of different places - namely in Functions and Operations
# We can also use them in Conditional Statements (IF Statements) as well

if job_work_from_home or job_health_insurance or not job_no_degree_mention:
  print('Meets Conditions')
else:
    print('Does not Meet Conditions')
```

### Membership Operators
- EG: `in, not in`
- They are checking weather an Object is inside of another Object

| Operator | Name           | Example    |
|----------|----------------|------------|
| in       | Membership     | x in y     |
| not in   | Not Membership | x not in y |

```python
# We want to check weather something is inside something else (Object inside another Object)
# We are checking to see if the word Data is inside the term Data Nerd and if so it will Return TRUE, otherwise FALSE
# Note this is case Sensitive... so if we looked for data, it would Return False

'Data' in 'Data Nerd'
```

```python
# From time to time it is used with Strings. But moreso it's used with things like Lists
# In our example we only have a List of 3 Skills, but imagine we have a List with hundreds of Skills in there, and we need to check it for a specific Value

job_skills_list = ['python', 'r', 'sql']

# Checking to see weather python is inside of the List
# It is so it will Return TRUE
'python' in job_skills_list
```

```python
# Conversely, we can check to see weather a Skill is NOT in a List
# It is in there so it will Return FALSE
# If python was not there and not inside the List, it would Return TRUE because it's checking to see if it's not in there
'python' not in job_skills_list
```

### Identity Operators
- EG: `is, is not`
- These are used to basically Identify weather something is correlated to another Item
- Now this commonly gets confused with the Comparison Operators

| Operator | Name         | Example    |
|----------|--------------|------------|
| is       | Identity     | x is y     |
| is not   | Not Identity | x is not y |

```python
salary_kelly = 100_000
salary_luke = 100_000

# In this case, we want to Compare the Values of these 2 Variables
# Doing this will Return TRUE
salary_kelly == salary_luke
```

```python
# Doing this will Return FALSE
salary_kelly is salary_luke

# This is because these Values are not Equal... So why is this?
# The Technical Reason: The IS Operator verifies if these 2 Variables are using the same Memory Location for the Value they are holding
# We can check this with the ID Function
# Below, we can see the Memory Locations are not the same and because of this they aren't Equal so IS will Return False, because they aren't the same Objects/aren't sharing the same Memory Location

print(id(salary_kelly))
print(id(salary_luke))
```

```python
# We can also check this using the following implementation:
# Because they aren't Equal, this Returns FALSE

id(salary_kelly) == id(salary_luke)
```

```python
# So when will using this Identity Operator in Data Analytics be relevant?
# Usually... it's not very relevant but it's good to know it exists and how to use it because it shows up from time to time

# A potential case of this is: We have a List of core Skills 
core_skills = ['python', 'r', 'sql']

# Then we said, these are Neil's Skills 
neils_skills = core_skills
```

```python
# In this case, we set them Equal to each other
# So now we can check to see if they are the same Object (occupy the same Memory Location)
# They are the same Object, so by using the IS Operator here, it will Return TRUE

core_skills is neils_skills
```
