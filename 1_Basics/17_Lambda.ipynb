# Lambda
___
- EG: `x, y: x + y`
- Lambda Functions, also known as Anonymous Functions are very similar in how they Operate compared to the User-Defined Functions adn also Built-In Functions
- In terms of Calculating and doing some soft of Operation
- Before we take a look at this, we have to be clear:
    - Lambda Functions are more of an Advanced Functionality
    - If you find yourself not using Lambda Functions and just using regular, User-Defined Functions there's nothing wrong with that
    - Lambda Functions are just a great way to Flex on your co-workers
- This one will be the hardest one and it will only get easier after this one!

```python
# Let's get into a Coding Example
# Lambda Functions are written by starting with Lambda!
# Then from there, you provide it an Argument, in this case we are going to use X (It can be any number of Arguments, but we are only going to do 1 for now)
# From there you do a colon to then go about providing what we want or what Formula we want to be conducting on that x Argument
# In this case, we are going to keep it simple for now we will have an x and multiply it times 2
# Lambda Syntax: Lambda Arguments is an Expression   

# The Printout states that it's just a Function here and it's telling us that we actually need to use it and see how it works
lambda x: x*2
```

```python
# We are going to name it and Run this
# Because it's now defined, we can use it and Call it (Python is the only language where you can Assign Functions to Variables)
mul_two = lambda x: x*2
```

```python
# Calling it as below:
# In Parenthesis, to call this Lambda Function (because it is a Function and we need Parenthesis to Call Functions)
# We can Provide it an Argument and get the Result

mul_two(3)
```

```python
# Now this is 2 Lines of Code right here, the point of a Lambda Function is to be able to write something within 1 Line
# We can now, take the Function and wrap it in Parenthesis and then put more Parenthesis on the outside of the Function to Call it and Provide the Argument for it

(lambda x: x*2)(4)
```

```python
# If we wanted to provide multiple Arguments to this Function
# We can add a comma and then we added y, then we need to give the condition for y (in this case we are making an addition), and the provide a Value as an Argument to pass it
(lambda x, y: x*2 + y)(4, 2)
```

```python
# So then... how does this Apply in the real world?
# To see this, let's go back to the previous example we had in the Functions Section
# Now with this Function, we have a List of Salary Values that we want to pass into it...
# And in our case, we are just going to assume that all the bonus_rates are in fact that 10% for all of these Salaries
# So how can we with what we know right now, actually implement this without using Lambdas? - Well we can use List Comprehension

salary_list = [100000, 200000, 150000, 12000, 80000, 750000]

# Calculate total_salary of base_salary * (1 + bonus_rate)
def calculate_salary(base_salary, bonus_rate=.1):
    return base_salary * (1 + bonus_rate)

# Now it will Run the Function on the salary_List and Print out the new Value of each Salary after iterating through it and Applying the Calculation
total_salary_list = [calculate_salary(salary) for salary in salary_list]

total_salary_list
```

```python
# But there's only 1 Problem with the above Solution... right now we are at 4 Lines of Code to Complete this
# We shortened it down by re-writing the Return Line and now the Function itself is at 1 Line of Code
# Any time we get a Function down to 1 Line of Code that's a use case for Lambda Functions!
# Now we are going to alter and Replace the Function with a Lambda Function as seen below:

# This is only an Example to show that we can write with Lambda Functions, one single line of Code that you can write things more summarized
# This is just to show how we can use Lambda inside of List Comprehension
total_salary_list = [(lambda x: x * 1.1)(salary) for salary in salary_list]

total_salary_list
```

```python
# Thing is, this isn't bullet proof and as summarized as it could be... and there are other ways to do this as well
# You could just do the below and we will get the same answer

total_salary_list = [salary * 1.1 for salary in salary_list]

total_salary_list
```

### Data FIltering Example
___
- Let's move into a slightly more complex Example of how we can use Lambda Functions in order to clean up or Filter our Data
    - Find jobs that have Python and that are Remote

```python
# Below we have a Dictionary of Jobs that are available, in there we have a Job Title, the different Job Skills and then also a Value on whether it's Remote or not as a Boolean Value
jobs_data = [
    {'job_title': 'Data Scientist',  'job_skills': ['Python', 'Machine Learning'], 'remote': True},
    {'job_title': 'Data Analyst',  'job_skills': ['Excel', 'SQL'], 'remote': False},
    {'job_title': 'Machine Learning Engineer', 'job_skills': ['Python', 'TensorFlow', 'Keras'], 'remote': True},
    {'job_title': 'Software Developer', 'job_skills': ['Java', 'C++'], 'remote': True},
    {'job_title': 'Data Scientist', 'job_skills': ['R', 'Statistics'], 'remote': False}
]
```

```python
# Because we want to Filter the Data, we are going to use a Built-In Function for this, known as Filter()
# So let's inspect it first!

help(filter)
```

```python
# Let's show what the Filter Function does very simpley
# We will define the filter function, followed by the function that we need to do, in our case Lambda. So this is where we define the Lambda Function
# And we are going to be Iterating through the jobs themselves, so we will Create a Variable called job 
# We are going to keep this simple right now, we just want to find those where the job of being Remote is True

# Running this, we get a Filter Object Returned and so we need to convert this to something we can actually see
# So we wrap it in a List Function to Convert it into a List
# We can see, we get back only the Jobs where Remote is True 
list(filter(lambda job: job['remote'], jobs_data))
```

```python
# But Remember, we don't want to just find where the Remote Jobs are True, we also want to Find where the job skills contains Python
# So we need to modify this and modify the Operation that we are doing in our Lambda
# We add in the AND Operator to make sure that both of these conditions are going to be True in order to satisfy the Filter Function
# We want to Verify that Python is IN the job_skills List of Values

list(filter(lambda job: job['remote'] and 'Python' in job['job_skills'], jobs_data))
```
