# Example: Analyze Salaries
___
- As we seen previously, using the `Group By` is a very powerful way of Aggregating and Analyzing different Groups of Data
- Now to do a deeper dive and Analyze the US/Canadian Job Market, specifically looking at the key statistics (median, min, max, count) around these different Job Titles available within their Country

```python
# Imports
import pandas as pd
from datasets import load_dataset

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
# First off, let's look at the Top 20 Value Counts for the Countries that are Available by looking at the job_country Column
df['job_country'].value_counts().head(20)
```

```python
# To find if a Country is anywhere in the Data, use:
df['job_country'].isin(['Antartica']).any()
```

```python
# Create a New Data Frame with only our Countries of choice Filtered for with that Job Data
us_jobs = df[df['job_country'] == 'United States']
canada_jobs = df[df['job_country'] == 'Canada']
```

```python
# Printing it out to Inspect the Data Frame
us_jobs
```

- Now we have the Data we need. We have 2 Data Frames, 1 for the US and 1 for Canada
- Now we can perform a Group By Aggregation in order to Analyze it for things like: **Median**, **Min**, **Max**, **Count**
    - Before we can do that though, in the salary_year_avg Column, there's a lot of NaN Values in there... these will need to be Filtered out first

```python
# Filtering out the NaN Values in salary_year_avg for us_jobs
us_jobs = us_jobs[us_jobs['salary_year_avg'].notna()]

# Filtering out the NaN Values in salary_year_avg for canada_jobs
canada_jobs = canada_jobs[canada_jobs['salary_year_avg'].notna()]
```

```python
# Printing it out to Inspect the Data Frame
us_jobs
```

- We have Successfully Filtered out the NaN Values
- Now we can do the Group By

```python
# Group the US Job Data
us_jobs.groupby('job_title_short')['salary_year_avg'].agg(['median', 'min', 'max', 'count']).sort_values(by='median', ascending=False)
```

```python
# Group the Canada Job Data
canada_jobs.groupby('job_title_short')['salary_year_avg'].agg(['median', 'min', 'max', 'count']).sort_values(by='median', ascending=False)
```
