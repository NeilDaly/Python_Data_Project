This Exercise is pretty simple:

- Filter what Jobs to Apply for based on a List of Skills.

```python
# We Define a List of Dictionaries
# Define data science job roles and required skills
job_roles = [
    {'role': 'Data Analyst', 'skills': ['Python', 'SQL', 'Excel']},
    {'role': 'Data Scientist', 'skills': ['Python', 'R', 'Machine Learning', 'Deep Learning']},
    {'role': 'Machine Learning Engineer', 'skills': ['Python', 'TensorFlow', 'PyTorch', 'Scikit-Learn']},
    {'role': 'Data Engineer', 'skills': ['Python', 'Apache Spark', 'Hadoop', 'SQL']},
    {'role': 'Business Intelligence Analyst', 'skills': ['Python', 'SQL', 'Tableau', 'Power BI', 'Excel']},
    {'role': 'Quantitative Analyst', 'skills': ['R', 'Python', 'MATLAB', 'Statistics']},
    {'role': 'Operations Analyst', 'skills': ['Python', 'SQL', 'Data Visualization', 'Process Improvement']},
    {'role': 'Database Administrator', 'skills': ['SQL', 'Oracle', 'MySQL', 'Database Management']},
    {'role': 'AI Engineer', 'skills': ['Python', 'TensorFlow', 'PyTorch', 'Computer Vision']},
    {'role': 'Statistician', 'skills': ['R', 'SAS', 'Python', 'Statistical Modeling']}
]

# My skills
my_skills = ['Python', 'SQL', 'Excel']
```

### Coding
___
- Loop through the `job_roles`
- Verify if all `my_skills` are inside each `job_roles['skills']`
- If `True`(or qualified), add to qualified list of jobs

```python
# Create an empty list where we will put the roles we qualify for qualified_roles.
qualified_roles = []

# Use a for loop to go through the jobs. And Set the qualified Flag to True
for qualified_jobs in job_roles:
    qualified = True
# Nested for loop to look through the skills for each role.
    for qualified_skills in my_skills:
        # An if statement checks if the skill in my Skill List is not in the skill list of the given role.
        if qualified_skills not in qualified_jobs['skills']:
            # If my skill is not present in the job postings skill list, set the qualified flag to False and check the next skill in my list until there are no more
            # If my skill is present in the job postings skill list, the we go to the next If Statement to Add the Job Role to the empty List
            qualified = False
            break
    # If it's true then add that role to the qualified_roles list we created.
    if qualified:
        qualified_roles.append(qualified_jobs['role'])
# Print out the Result
qualified_roles

# Uncomment the below to see how the Code iterates through and what is stored where. You can uncomment to understand what is happening
#qualified_jobs
```

```python
# Additionally, this could have been done using the ALL Function...
# What it does is using a List Comprehension inside of it to comapre the Skills and make sure the my_skills List is inside the job skills

# Create a Blank List so we can determine which jobs you're qualified for
qualified_roles = []

for job in job_roles:
    # Check if all required skills are in my_skills
    if all(skill in job['skills'] for skill in my_skills):
        # Add the job role to the qualified_roles List
        qualified_roles.append(job['role']) 

qualified_roles
```

```python
# Because the above is so short, we can make it into a List Comprehension

qualified_roles = [job['role'] for job in job_roles if all(skill in job['skills'] for skill in my_skills)]

qualified_roles
```
