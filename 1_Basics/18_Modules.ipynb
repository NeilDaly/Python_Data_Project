# Modules
___
- EG: `import module`
- We've have been Focusing a lot on Functions for the last while
- Say now we want to move onto a New Project or New Python File... and we want to write a Function in there, but we already have used/written that Function!
    - Would we have to Re-Write it again to use it? That's where Modules come into play
- Continuing on with Functions, we are moving more into getting other Functions or even Classes instead of Building our own Functions
- This is done through things, like Modules (which is the Focus for this Lesson) and even in the next Lesson when we switch Focus to Libraries! (Libraries are just a collection of Modules)

```python
# Now let's get into Building our own Module and we will do it Step by Step with a simple Example
# In the left Menu, you can see we Created a New .py File (Python File) named my_module - Modules are typically named all lowercase
# So now that we have Defined a Skill List inside the Module, how do we get it? How do we Access it?? Well we now have to Import it as seen below

import my_module

# Now we want to Call the Skill List that is inside the module - We want to see it in here
# We can do this by using dot notation and accessing it like we do with Methods

my_module.skill_list

# Bare in mind, Listing Variables inside of a Module is not a common use case
# What is a common use case? Defining Functions inside of a Module!
# So now, let's do that! Because we now have it Created and Defined, let's Call it!
# The Synatax to Call a Module's Function is: module_name.function_name()
# When trying to Run this, we get an Error... why?
# Well we Saved the File and Updated it with the Function inside of it, but the Problem is the Updated Contents whenever we Imported it in are not Updated inside of this Environment
# It's imported in and Saved into Memory in the initial state it was Imported in at... because we made changes AFTER Import, our changes/Updates are not in there, and now it's trying to Find and Call a Function that doesn't exist because it's Referencing the File in Memory, rather than the live File
# It's really lazy, even though we Saved it and Re-Ran the Import, it doesn't go through and actually grab the New Contents of the File
# It keeps the Initial File Loaded into Memory and References that instead of fetching the newer Version... so we basically have to start the whole Session all over again...
# To do this, we go to the top of the IDE and Select, 'Restart', then select 'Run All'  - Now the Function Runs and Executes Returning what we Expect!
# Just like with an Electronic Device where you have to turn it off and turn it back on again, sometimes if you Run into a problem with Jupyter Notebooks, just flip it on/off

my_module.favorite_skill("Python")
```

- Now let's get into a Practical Example of how you'd actually set this up for a real case Scenario where we will need to use Modules
- Once again, we will Create a Module inside of our Files by Creating a New File inside the Folder and naming it something appropriate to what it actually does!
- In here we will be Creating a lot of different Functions where we will Analyze a lot of different Job Data so, we will name it: job_analyzer.py

```python
# Import the Module
import job_analyzer

# Call the Function inside there
job_analyzer.calculate_salary(100_000)
```

- It's important to Note, that Import isn't the only Statement that we can use in here
- It's very cumbersome to write out the Module along with the Function Name after it...
- If we are only using one or two Functions, there's an easier way to do this!
- Instead of only using that `import` Keyword we can use another Keyword and that is `from` along with specifying what exactly you want to Import
- We can use these in conjunction with each other! See below:

```python
# This is stating from the Module File, you want this specific Function
from job_analyzer import calculate_salary, calculate_bonus

# Call the Function as normal and it will Run and Execute as expected!
print(calculate_salary(100_000))
print(calculate_bonus(110_000, 100_000))
```

- Now let's say we have Multiple Functions inside of our job_analyzer Module (I just Added a second one in there)
- And now we want to Import both of these in
    - Instead of doing `from job_analyzer import calculate_salary, calculate_bonus` (which would work just fine as you can see above)
- This is cumbersome to do it this way because if we have mutiple different Functions, we would have to List them out one by one and then it gets hard to read and becomes a long list
- If there's not a lot of Functions, inside of the Module itself, we can do it this way `import *`
    - Like in SQL, the Star means all, so we Import **ALL** - It will Import all the different Functions that are defined inside of there as seen below:

```python
# Run the Imports with the Star Notation which gets ALL
from job_analyzer import *

# Call our Functions and provide Arguments as normal - Wrapping them in a Print Function as well so I can see both Outputs, they Run and Execute as Expected
print(calculate_salary(100_000))
print(calculate_bonus(110_000, 100_000))
```

```python
# We can also Run the Help() Function on the Functions contained inside of our Module
# This is particularly helpful, especially when we are giving our Module File to our co-workers so they can use it as a Module File as well
# Helpful, especially when they are trying to Run and use this and they may have questions with it and what's going on in here
# Now Running this after adding in the Doc String, we can see the Comment Prints out alongside the Function from the Module File after Running the Help() Function on it

help(calculate_salary)
```

```python
help(calculate_bonus)
```

# Python Standard Library

### Statistics Module

```python
# List of Salaries
salary_list = [98_000, 101_000, 102_000, 99_000, 97_000]
```

```python
# Import the Module: Statistics
import statistics

# Using the Mean Function that comes from the Statistics Import on the Salary List 
statistics.mean(salary_list)
```

- Now what happens when we want to use Mean, Mode and Median?
- That will be cumbersome to write out all that so we can just Import them directly so we can use them
- As a Note: It's better to List out the Imports that you want rather than using the `*` to bring them all in
    - This is because as the Module List gets longer and we Import more, we don't want to Import all using the `*` because then we will Import stuff we aren't using
    - Think of if we have 5 Modules Imported, and we did `*` for all those, but we only used 6-7 of the Functions in those Modules, then it's a waste bringing them all in
    - As well as this when we start using bigger Modules and more of them, it's better to List them out so we can keep proper track of which ones we are using and which ones are relevant to our Project
- Important to Note, as well, if we use the `*` and bring in everything, if we name Variables that are Keywords in the Modules, we will Run into Issues and Errors Running the Code because the Variable will have a Keyword but we won't know where the Error is because we have Imported everything 

```python
from statistics import mean, mode, median

print(f'The Mean is:', mean(salary_list), f'The Mode is:', mode(salary_list), f'The Median is:', median(salary_list))
```
