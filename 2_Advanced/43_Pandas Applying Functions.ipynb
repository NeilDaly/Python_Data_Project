# Pandas - Applying Functions

- EG: `df.apply(lambda x: x * 2)`
- We haven't even jumped into Exploring and Analyzing this `job_skills` Column that has that List of Skills and that's because there's a slight problem with this
- It's not contained as a List inside of our DataFrame right now. It's being Annotated as a String...

- But before that, we will start with some simple examples first!

```python
# Imports
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Load Dataset
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

- For the first and second Example, we are going to be looking at the `salary_year_avg` Column
- And there's a bunch of NaN Values in there so we are going to go ahead and Filter those out first using `.notna()` Pandas Method
    - We go ahead and specify the `df`
    - In here, we Run the **Pandas** Method of notna `pd.notna()`
    - We specify where we want to Filter the NA's from, so the `df` and the Column Name, `df['salary_year_avg']`
    - Then adding in the Column Name `['salary_year_avg']` at the end so we only Return that Column, otherwise, we Return the entire df with the Filtered Results

```python
# Filtering out NaN Values from salary_year_avg
df[pd.notna(df['salary_year_avg'])]['salary_year_avg']
```

- Now we have this back, this is the Salaries that we have
- What we want to do with this Problem is
    - We want to Calculate the Projected Salary Next Year
    - What we can do is just some rough Maths as an Example. We will Apply what Inflation is right now of around 3%
    - So basically, take the Salary Values and Multiply it times 1.03. Since the Inflation is 3% (0.3) and it will go up by that, so we need the 1 to Account for the now Values
    - Otherwise, if we only do 0.3, we will need to Add that Result to the now Values to get the Total, so might as well just lump it in there from the beginning
    - This will get to see what will be 3% higher next year
- With that, how can we actually get, inside the DataFrame how can we get it to **Apply** to this entire Column of `salary_year_avg`
    - Well! There's a Function for that!
    - `pandas.DataFrame.Apply()`
    - This can be Applied on a DataFrame, it's a Method of a DataFrame
    - Takes 1 or Multiple Arguments and we are only going to focus on the first one for the time being
        - func: It's a Function. A Function to Apply to each Column or Row
        - We will get into how we can Apply this to Rows, but for now, we will focus on 1 Column

```python
# Running the Help Function on the Apply Method
help(df.apply)
```

- First thing we need to do, is Create a Function!
    - Start off with the `def` keyword and give it a Name of `projected_salary()`
    - It's only going to take 1 Argument of `(salary)`
    - With that Salary, what we want to `return` out of this Function is the `salary * 1.03` (so that 3% Inflation)
- From there, it's pretty simple!
- We can now take a Column of interest from our DataFrame such as `df['salary_year_avg']`
    - And use the `.apply()` on it
    - Providing the Function we Created of `projected_salary`
    - Because we are providing the Function, we don't need to provide the Parenthesis `()` after the Function. We are just Providing the Function Name, it will Apply it as necessary
- Running this on the Altered DataFrame where we Removed the NaN Values  

```python
# Creating a New DataFrame called df_salary and setting it equal to the Filtered DataFrame with the Removed NaN Values
df_salary = df[pd.notna(df['salary_year_avg'])].copy()

# Creating a New Function so we can Run the apply() Method on it and get back the Results
def projected_salary(salary):
    return salary * 1.03

# Taking our Column of intrest, salary_year_avg, from our new Altered DataFrame that Removes NaN Values and Running .apply, providing the Created Function
df_salary['salary_year_avg'].apply(projected_salary)
```

- Now we have back all those different Salary Values!
- Now, let's actually inspect this and see how it Compares to that Original `salary_year_avg` Column
- So, we will Create a New Column inside our Filtered DataFrame called `salary_year_inflated` and Set it Equal to what we did above with Multiplying `salary_year_avg` by the Created Function
- And then for the New Filtered DataFrame, I want to show, these 2 Columns of `salary_year_avg` and `salary_year_inflated`
    - Doing this, we can Compare the Values in the 2 Columns against each other
    - Easier to see when we Compare the Original Value to the Inflated Value
    - Now we can see the Projected Value as it Compares to the Original Value

```python
# Creating a New Column in the DataFrame to Compare the Values against
df_salary['salary_year_inflated'] = df_salary['salary_year_avg'].apply(projected_salary)

# Printing out the 2 Columns from the New DataFrame to Inspect them and Compare Values
df_salary[['salary_year_avg', 'salary_year_inflated']]
```

- Inspecting this further, we can see that `salary_year_avg` Column was Multiplied times 1.03
- Looking at the Function, we can see that it's pretty short... and Providing the `.apply()` Method this Created Function, we could use something like an **Anonymous Function**
    - So let's try writing this again with an **Anonymous Function**
        - Declaring our New Column Name of `salary_year_inflated` and Creating it in our `df_salary`with this notation: `df_salary[salary_year_inflated] = `
        - Define `df_salary` again so we can assign what we are going to do to the Old Column, in this case, manipulate the `salary_year_avg` Value and store it in our New Column
        - The notation will be `df_salary['salary_year_avg']`
        - Then we will Apply a Lambda Funtion to it using `.apply()`
            - For this because we haven't used Lambda Functions for a while... so we will need a Reminder!
            - Define `lambda`
            - Then Define a specific Variable, so in this case, we will define `salary`
            - Then you will need to Add a Colon `:` so it will look like this `salary:`
            - After the Colon, you will need to provide whatever Mathematical Equation/Operation you want it to do! In our case, we want `salary * 1.03`
    - Then once again we will want to see, what the `salary_year_inflated` Column looks like, compared to the `salary_year_avg` Column so we will want to Print both of those out
    - Notation is `df_salary[['salary_year_avg', 'salary_year_inflated']]`

```python
# Doing the exact same Function but this time doing it via an Anonymous Function
df_salary['salary_year_inflated'] = df_salary['salary_year_avg'].apply(lambda salary: salary * 1.03)

# Printing out the 2 COlumns for Comparison
df_salary[['salary_year_avg', 'salary_year_inflated']]
```

- Once again, comparing it to the last one, we get the same Results, which means all the Math was done correctly with the Anonymous Function
- For the hard-core Data Nerds, you may have noticed, that we can re-write this without the `.apply()` Method or Lamda Function!
    - Since, all we are actually doing is, Multiplying 1.03 by another Number in the `salary_year_avg` Column
    - And the Printing out the 2 different Columns to see what the difference is
    - Since this is an overly simplistic example, we don't have to use any of these Methods or Functions, we can just do the Maths on this explicitly
    - See below what I mean!
- But this was mainly to show an Introduction to the `.apply()` Method when Running it on a Column!

```python
# Doing the Math Raw and Explicitly, without Lambda or Apply Functions
df_salary['salary_year_inflated'] = df_salary['salary_year_avg'] * 1.03

# Printing them both out
df_salary[['salary_year_avg', 'salary_year_inflated']]
```

    - That's in the context of our `job_skills` Column
    - Specifically, we need to Convert this from a String into a List
- If we wanted to Display the List of Skills from the second Record of our DataFrame, we can do it as such

```python
# Displaying the List of Skills from the second Record of the DataFrame
df['job_skills'][1]
```

- After Running, we can see that the List has double Quotes around it...
- And even wrapping it with a `type()` Method, we can Confirm that it is in fact, a String

```python
# Running Type() to Confirm it's DataType
type(df['job_skills'][1])
```

- So how can we actually Convert this back to a List?
- You could wrap certain Objects inside of a DataType like List and then try to Convert it...
- However, in this case, if we were to try this we get....

```python
# Converting the String Type List, into a List
list(df['job_skills'][1])
```

- This absolute jumble of letters back...
- It turns it into a jumbled mess that is not possible to read...
- It Converts every single Letter into it's own Item, into a List
- Basically it doesn't work!

- Instead, we are going to use the `ast` Module! Which stands for **Abstract Syntax Trees**
- This is special for our Unique use case and not something that would be common
    - `ast.literal_eval()`
        - With this, you provide a Node or a String (In our case, we are going to Provide that String to it)
        - Basically goes through and it Converts it to the Container DataType, in our case a List - Back to what it's supposed to be!
- Let's actually show this in action!
    - First we will need to Import in the `ast` Module and it's apart of the Python Standard Library so no Install required - `Import ast`
    - Then we are going to Apply the `literal_eval()` Function on the `df` for the String List Column of `job_skill` and only for the Second Record to see it working

```python
# Importing in the Module
import ast

# Running literal_eval() on the String List Column to Convert it to a List only! Doing it for 1 Record first to see it working
ast.literal_eval(df['job_skills'][1])
```

- Now we have this back into a List
- We can also Confirm this by Running `type()` on this as well and seeing it

```python
# Running type on this
type(ast.literal_eval(df['job_skills'][1]))
```

- It comes back as a List and Confirms that the Conversion was Successful!
- So let's now clean up this Column using the `Apply` Function
- Just as a Note as well! We can't actually Run `ast.literal_eval()` on this entire Column. We are unable to explicitly do this
- If we try and do this, we will get a Value Error as we didn't pass in the stuff correctly, we **have** to use the `apply()` Method for this

```python
# Trying literal_eval() explicitly on the Column
df['job_skills'] = ast.literal_eval(df['job_skills'])
```

- First thing we are going to do is to Create a Function
- Using the `def` keyword, and naming it `clean_list()`
- We will need to Pass in a Variable as well, so we will call it `skill_list`
    - Notation so far is `def clean_list(skill_list):`
- This Function is going to be going through each of those Items in the Column so we can Return that Literal Eval of our Skill List like so
    - Notation is `return ast.literal_eval(skill_list)` 
- Now that the Function is done, let's write out our `apply()` Function
    - For this, we just want to Replace it, right on top of the Original Column of `job_skills` so we will set it equal to itself
    - Then using the `.apply()` Method we will pass in our Created Function
    - Notation will be `.apply(clean_list)`
- Running this, will actually give us an Error!
    - Specifically, we get a **ValueError**
    - This is because of a Malformed Node or String... it's basically because we have NaN Values in our String List

```python
# Writing our Function to Convert the String List into a List and Return only the List
def clean_list(skill_list):
    return ast.literal_eval(skill_list)

# Using the Apply Function to grab our Created clean_list Function in order to Clean the List
df['job_skills'] = df['job_skills'].apply(clean_list)
```

- We can show this by Filtering the DataFrame to only show NA Values inside of the `job_skills` Column
- We can do this by using `.isna()` Function and passing in what Column we want to Filter it for

```python
# Using isna Function to see the Jobs that don't have Skills Listed
df[pd.isna(df['job_skills'])]
```

- Inspecting the Results and looking at our `job_skills` Column, we can see that we have a bunch of NaN Values
- This is causing Errors for us here, because it's trying to do the `literal_eval` on NaN when it's expecting a String to be Passed
- How do we fix it?
    - We can Pass in an `if` Statement inside of our Created Function which checks if the Value is `notna()` then do the `literal_eval` and Return the Manipulated Value
    - If it is NaN, we do nothing and skip it leaving it as is
    - Using the `pd.notna()` Function and passing in our `skill_list`
    - If it's not NA (has a Value) then we Return it as a List otherwise, we are going to leave it None

```python
# Writing the Function with an if Statement inside of it to check for non NaN Values
def clean_list(skill_list):
    if pd.notna(skill_list):
        return ast.literal_eval(skill_list)
    
# Using the Apply Function to grab our Created clean_list Function in order to Clean the List
df['job_skills'] = df['job_skills'].apply(clean_list)
```

- Everything Ran right and there were no errors, which indicates Success!
- But let's actually Inspect the second element of this to Confirm if the `job_skills` are now in a List

```python
# Inspecting the second record to make sure it's in a List
df['job_skills'][1]
```

- Printing it out, it is now in a List as we got a List back!
- Confirming the Type of this by Running the `type()` Function on this

```python
# Confirming Typing
type(df['job_skills'][1])
```

- We can see that it is infact a List!
- This Confirms, that it did do the Conversion correctly!
- In this case, because of how simple this Function is... it's encouraged, that we make this into a `lambda` Function
    - It's important to Note, it's good to write a Function first, and then Convert it afterwards!
- Starting this off the same way, by overwriting the Original `job_skills` Column with what we do to it. To do this, like always, we make it equals itself 
    - `df['job_skills'] = df['job_skills']`
- We then Invoke the apply as well which again is the same 
    - `.apply()`
- Inside the `apply()` brackets is where we do our `lambda` action
    - Start be defining `lambda` and define the Variable that we want to use, we will call it `clean_skills`
    - Then it will be followed by a Colon as well! So it will look like `lambda clean_skills:`
    - Now we get into what we actually want to do, we want to perform the converstion on the Variable and change it from a String List to just a List
    - We do an `ast.literal_eval()` to accomplish this and we pass in the `lambda` Variable we defined like this - `ast.literal_eval(clean_skills)`
    - Running just that, we will get the same Error as before, complaining about NaN Values, so we will need to Filter those out using an `if` Statement
    - The `if` Statement needs to be included inside of our `lambda`. It will be same as the one we Created above. We need to put it **after** the `literal_eval` Statement
        - Since it's the same it will be `if pd.notna(clean_skills)`
    - Running just this, we will also get an Error!
        - This is because it won't know what to do with the NaN Values and it expected an `else` after the `if`
    - Adding in the `else` to just Return back that `clean_skills` List if it is None/NaN Value (what the Value is currently)

```python
# Doing the exact same Function but this time doing it via an Anonymous/lambda Function
df['job_skills'] = df['job_skills'].apply(lambda clean_skills: ast.literal_eval(clean_skills) if pd.notna(clean_skills) else clean_skills)
```

- Running this, everything looks like it Ran fine
    - But we can check and Confirm with the `type` Function to make sure the Records are of Type List

```python
# Confirming it's a List with the type Function
type(df['job_skills'][1])
```

- It is in fact a List!

- Previously, we only covered the Function aspect of it, but now we need to look at the `axis`
    - Default currently is 0, which means that it applied the Function to each Column...
    - But if we want to Apply a Function to a Row, so we can access multiple Columns in a Row or check a Column this is what we want to use
- Because this one is a little bit more Complex, we are going to be building a Function first, and then Passing it into our `apply()` Method
- Let's start by Building out, how we are going to use that `apply()` Method
    - For this, we are going to be using the `df_salary` DataFrame again because it already has all of the Salary Data Filtered out and makes it easier to Display
    - First we are going to be Creating a New Column of `salary_year_inflated` and in this case, instead of providing the DataFrame and the Column we are just going to provide the DataFrame
        - Because we want the entire Row to go to it this time
        - We are going to Run the `apply()` Method on that
        - We are going to be using a Function (that is pending Creation at this point) but it will be called `projected_salary`
        - This will be passed into the `apply()`
        - Specify the Parameter of `axis=1`
- Now it's time to start building out the Function
    - Starting off as usual with the `def` Keyword, the name of `projected_salary()`
    - Inside of it, we will be Passing the Row so we call the Variable, `row`
        - What we need to do inside of this as it's grabbing each one of these Rows is check what the Value is in the `job_title_short` Column
            - If it's **not** Senior, take the `salary_year_avg` and Multiply it by 1.03 
            - If it **is** Senior, take the `salary_year_avg` and Multiply it by 1.05
            - This is a great `if` Statement
            - It will look like this, `if "Senior" in row['job_title_short']:` `return 1.05 * row['salary_year_avg']`
            - Then we will need an `else` Statement to handle all the other Values
            - It will look like this, `else: return 1.03 * row['salary_year_avg']`
- For Printing, we want to Display, the `job_title_short`, `salary_year_avg` and the `salary_year_inflated`
    - To do this, we do the usual `df_salary[['job_title_short', 'salary_year_avg', 'salary_year_inflated']]`

```python
# Building the row Example
# Function that Calculates: Senior Roles are going to have an Inflation/Projected Salary of 5% higher and the other Roles are only 3%
def projected_salary(row):
    if "Senior" in row['job_title_short']:
        return 1.05 * row['salary_year_avg']
    else:
        return 1.03 * row['salary_year_avg']

# Apply Function, we want to apply the entire DataFrame, so as to give the Function access to every Column in the whole DataFrame and axis=1 so it Applies to the Rows instead of Columns 
df_salary['salary_year_inflated'] = df_salary.apply(projected_salary, axis=1)

# Printing out to see the Results
df_salary[['job_title_short', 'salary_year_avg', 'salary_year_inflated']].head(17)
```

- Running this, all the Values are going to Fill in depending on if it's a Senior Role or not, and we can see that it worked correctly
- If it's a Senior Role, it's doing the 1.05 Multiplication Formula and for everything that's **not** a Senior Role, we can see it did the 1.03 Multiplication Formula
- The above Code, can actually be Transformed and written into a `lambda` Function as well
    - This will simplify it into one line... buuuut it does start to look really long but it can be done
- See below, the same Function as above, doing the exact same Calculation but in the form of a `lambda` Function

```python
# Doing the same Function as above, but as a Lambda Function
df_salary['salary_year_inflated'] = df_salary.apply(lambda row: 1.05 * row['salary_year_avg'] if "Senior" in row['job_title_short'] else 1.03 * row['salary_year_avg'], axis=1)

# Printing it out
df_salary[['job_title_short', 'salary_year_avg', 'salary_year_inflated']].head(17)
```
