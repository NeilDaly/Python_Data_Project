```python
# Imports
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading the Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
# Create a New Column that will Contain only the Shorthand Month Value
df['job_posted_month'] = df['job_posted_date'].dt.strftime('%b')
```

```python
# Creating an Array of all the Different Months
months = df['job_posted_month'].unique()
```

```python
# Creating a Dictionary Comprehension
dict_months = {month: df[df['job_posted_month'] == month] for month in months}
```

```python
# Running the concat Function to combine all this Data
df_q1 = pd.concat([dict_months['Jan'], dict_months['Feb'], dict_months['Mar']], ignore_index=True)
```

```python
# Printing out our DataFrame
df_q1
```

- Moving into the next one, `.to_csv`
    - Used when we are frequently doing some sort of Analysis and Creating different DataFrames. We want to Export in and Save that DataFrame in case we want to Access it and Analyze it in the future
- Inside of `.to_csv`, we are going to specify the File Name that we want it to Save to
- After Running, we can see that the File has been Saved and it's in the same Folder as where I am Saving all these Files

```python
# Moving onto the .to_csv Method
df_q1.to_csv('quarter_1')
```

- Inspecting it, we can see, it is Saved as a CSV, but it doesn't have the correct File Extension appended to it
- Whenever you Run this, you need to Run it with the File Extention that you want to Save it as 
    - In this case, we want to Save it to a .CSV

```python
# Running it with the Extension Added
df_q1.to_csv('quarter_1.csv')
```

- We can see that it was Saved into the Folder with the correct File Extension Appended onto it
- Inspecting it, it looks like it has all the Data. Scrolling to the bottom we can Verify it has the same Number of Rows in there as our DataFrame
    - This Proves, that the Export worked and included everything in there that we wanted
- Now, if we ever wanted to Read that File
    - All we'd need to do is Run `pd.read_csv('quarter_1.csv')`
    - Just need to Invoke the `.read_csv` Method and Specify the File Name along with Extension, we would want to Read

```python
# Reading the CSV
pd.read_csv('quarter_1.csv')
```

- Inspecting it, looks like we have all of our different Data, 220K Rows
- One thing to Note with this is, the First Column we see after Reading the File
- It is called **Unamed:0**
    - We didn't specify, whenever we looked at this, that the first Column is the Index
    - Also, in the File, we never gave the Index a Name, so looking at the File there's nothing in there for the Name of the Index, it's a Blank Column Name
- What we can do when we Read the CSV is specify the Index Column. We can do this by
    - `index_col=0`
- Running this fixes that Issue

```python
# Fixing the Index Issue
pd.read_csv('quarter_1.csv', index_col=0)
```

- Let's now Export this Data into the same File Format that our Boss wants
    - Specifically into Excel
- We will use `.to_excel` for this
- Similar to `.to_csv`, we will need to specify a File Name that we want to Save it under and also specify the File Extension that we want to Save it under as well

```python
# Moving onto the .to_excel Method
df_q1.to_excel('quarter_1.xlsx')
```

- NOTE: Excel has a 1 million Row Limit so that is something to consider, if the Data is Greater than 1 Million Rows

- The 3 other Methods that we need to be aware of when we become more Advanced at Python are:
1. `.to_sql`
- For this one, we would need to Import a Library, like `sqlalchemy` and then from there, we can Export it to a Database of our choosing
- If we have a Connection to it, we could Set that up as well and then Export it Directly to that
___
- These next 2 are going to Save us some time if we are dealing with really large Data Sets and we are trying to Export them and also Load them into Python very quickly
- These 2 File Types are very efficient Files. So whenever these are used, you can then Read in the Data alot quicker
- Sometimes, even faster than CSV's
2. `.to_pickle`
3. `.to_parquet`
