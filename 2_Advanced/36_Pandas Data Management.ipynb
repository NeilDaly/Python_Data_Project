    - `df.head()` - Looks at the First 5 Records
    - `df.tail()` - Looks at the Last 5 Records

- Luckily, in cases like this we do in fact, have a Method for that!
- Similar to the `head()` and `tail()` Methods, we have the `sample()` Method
    - `pandas.DataFrame.sample`
        - This Returns a Random Sample of Items from an Axis of Object
        - Just like `head()` and `tail()`, we can specify the number to Sample
        - Or specify the fraction of Float. How much of the Fraction of Data to Return

        - Something to Note, everytime this is Ran, it will provide new/different Results within this Sample

        - This is seen further down (Keep Scrolling)
        - Now when Running this, the same Results are going to Return every. single. time.
        - Now, we don't have to worry about any Variation in what we are Returning back

```python
# Sample Method
df.sample(10)
```

```python
# Sample Method using the random_state Argument
df.sample(10, random_state=42)
```

- Next Method we are going to cover is `copy()`
    - `pandas.DataFrame.Copy`
        - Make a Copy of this Object's Indices and Data
- Let's do some Practice with this Method
    - Let's Create a New Data Frame called `df_original` and all we are doing is Loading our Dataset into it

```python
# Creating a New Data Frame and Loading our initial Data Frame into it
dataset = load_dataset('lukebarousse/data_jobs')
df_original = dataset['train'].to_pandas()
```

```python
# Inspecting our Data Frame to Confirm everything is in there and it's the same as the Original
df_original
```

- As expected from the name of the Method, the `df.copy()` Method Creates a Copy of the Data Frame
- Now, you may be wondering, why do we need to use this `df.copy()` Method?
    - Because usually all we need to do is this: `df_altered = df_original` (where we Create this new Data Frame, `df_altered` and it's equal to the Original Data Frame, `df_original`)
    - And then if we were to Print out the `df_altered`, everything would be in there and we'd get a perfect Copy back
- So what's the purpose of this whole `df.copy()` Method?
- Well... let's look at this, using an Example
- But as a short Explanation:
    - When we do the Assigning, the Unique ID assigned to the Data Frame, is also given to the New Data Frame
    - So any changes we do to the Altered, will be done to the Original, since they will have the same ID's
    - This is bad, because when we do multiple different Calculations to Multiple different Data Frames, we don't want them Altered, because what's the point in having different Data Frames?
    - Now all the Data gets compromised and it becomes a mess
    - That's why when we Create a Copy of the Data Frame, it will Create a True Data Copy, but will Assign it a New ID and leave the Original intact
    Quick EG:
    1. `df_altered = df_original`
    2. `df_altered` has ID 12345 
    3. `df_original` also now has ID 12345
    4. We do an override of NaN Values for `df_altered`. Where we Replace the NaN Values with the Median Values
    5. Expected Outcome: `df_altered` has the NaN Values Replaced with Median Values and `df_original` remains intact and is untouched by the Operation
    6. Actual Outcome: Not only has `df_altered` NaN Values been Replaced with Median Values but `df_original`'s NaN Values have also been Replaced by Median Values
    7. This is because they share the same ID or Unique Identifier - different to Memory Location
    8. To Prevent this, we use the `df.copy()` Method so the Data is a Perfect Copy, but the Unique ID for the Data Frame is different, thus any alterations are only made to the New Data Frame, leaving the Original untouched and intact

```python
# Assigning our Original Data Frame into a New Altered Data Frame Variable
df_altered = df_original
```

```python
# Looking at the Previous Example where we Replaced the NaN Values with the Median salary_year_avg Value
# Inspecting this Altered Column, we can see most of them have NaN Values (All of the shown 5 have NaN Values)
df_altered.loc[:5, 'salary_year_avg']
```

```python
# Calculate the Median Salary
median_salary = df_altered['salary_year_avg'].median()
# Printing it out
median_salary
```

```python
# Using the Altered Data Frame, Fill in those NaN Values with the Median Salary Value we Calculated
df_altered['salary_year_avg'] = df_altered['salary_year_avg'].fillna(median_salary)

# Print out those Values to show our work
df_altered.loc[:5, 'salary_year_avg']
```

- Now we can see those NaN Values previously are now Filled in with that Median Value of 115,000
- But... what about our Original Data Frame? -> `df_original`
- Those should still be Nan Values.... right? Since we didn't touch those!
    - Well actually... as you can see below, they are not... and they are now Updated
- When Inspecting it, we actually find, that the Original Data Frame actually got Updated just as our Altered Data Frame did

```python
# Inspecting the Original Data Frame
df_original.loc[:5, 'salary_year_avg']
```

- This is not a good thing! Because:
    - Technically, we can have multiple different Data Frames that have done multiple different Calculations on and we wouldn't want those Calculations on 1 Data Frame to affect the other
    - So... why is this happening?
- Well, it has to do with the fact of how we Assigned `df_altered = df_original`
    - When we used this Assignment Operator (`=`), the `df_original` had a Data Frame in it that had a Unique ID
    - This new `df_altered` was also set equal to that same Data Frame
- That is all a bunch of Theory, so let's go and prove it!

```python
# Getting the Unique ID of the Original Data Frame
id(df_original)
```

```python
# Getting the Unique ID of the Altered Data Frame
id(df_altered)
```

- Comparing the 2 Results. The Numbers are the same!
- Looking at the Returned Values we can see that these Numbers are identical and this proves they are sharing the same Unique ID

```python
# Because we are too lazy to Compare Number by Number to Prove they are the same
# We can just use the Comparison Operator to do this for us
id(df_original) == id(df_altered)
```

- So, these Variables are Referencing the same Data Frame... which is very not good!
- Now, let's show this in action
    - Let's Reload the Data Set so we have this Unique ID for our `df_original`

```python
dataset = load_dataset('lukebarousse/data_jobs')
df_original = dataset['train'].to_pandas()
```

- Now let's Create that `df_altered` again and we will set it equal to `df_original`
- But this time, let's use the `df.copy()` Method alongside it

```python
# Creating df_altered and setting it equal to df_original but invoking our Copy Method
df_altered = df_original.copy()
```

```python
# Fancy Print Statements like before where we show the ID's of: Original, Altered and Comparing the two
print('ID of df_original:                ', id(df_original))
print('ID of df_altered:                 ', id(df_altered))
print('Are the two Data Frames the same? ', id(df_original) == id(df_altered))
```

- With this, we can see that the `df_altered` Data Frame is different
- And it also Confirms this with the Comparison Operator saying that it is False
- And now we Calculate that Median Salary that will Replace the NaN Values in the `salary_year_avg` Column

```python
# Calculating the Median Salary
median_salary = df_altered['salary_year_avg'].median()
```

- Replace the Nan Values with the Calculated Median Salary Value

```python
# Replacing NaN Values with the Median Salary Value
df_altered['salary_year_avg'] = df_altered['salary_year_avg'].fillna(median_salary)

# Printing out those Values to show our work
df_altered.loc[:5, 'salary_year_avg']
```

- It's done correctly on `df_altered` as we can see

```python
# Printing out those Values to Prove that the Original is untouched
df_original.loc[:5, 'salary_year_avg']
```

- For `df_original`, this one is **unaltered**
