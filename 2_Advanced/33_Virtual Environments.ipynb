# Virtual Environments
___
- This next Section we will Create a New Virtual Environment specifically for the Project in this Course
- This is important for a few Reasons:
    - Helps with Version Control
    - Don't have to worry about Version Conflicts
    - Better off doing this than using the Base Environment
        - While the Base Environment is easier to use for quick one off scripts, it can easily become cluttered and prone to Version Conflicts while working on different Projects
    - **Dependency Isolation**: Different Projects often Require different Versions of the same Library (e.g., Project A needs Pandas 1.5, while Project B needs Pandas 2.0).
        - Environments keep these separate so one Project doesn't break another
    - **System Protection**: Modifying the Base (System), Python can accidentally break OS-level Tools that rely on specific Python versions.
    - **Reproducibility**: You can easily Generate a list of a Project's exact Requirements (e.g., requirements.txt) so others can run your Code without Troubleshooting Installation Errors. 
    - **Clean Workspace**: It prevents your Global Python Installation from becoming a "yarnball" of old and unused Libraries.

- To do this, we will need a few Commands
- This will all be done in the Terminal as well which can be accessed inside VS Code
    - Shortcut: CTRL + SHIFT + ~
- Once in there, we can start running the Commands

```python
# To check what is Installed

conda env list
```

```python
# To Create a New Virtual Environment

conda create -n test python=3.11
```

```python
# To Activate the Environment

conda activate test
```

```python
# To Deactivate the Environment

conda deactivate
```

```python
# To Delete the Environment

conda remove -n test --all
```

```python
# To Install the Kernal 

conda install ipykernel
```

```python
# To Install Packages needed/Libraries

conda install pandas datasets matplotlib
```
