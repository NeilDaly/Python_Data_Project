### Back to Learning Pandas and Learning Python
- Let's set up the Imports and Data Loading as we have previously done before

```python
# Imports
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
# To look at the first row of Data we can use
df.iloc[0]
```

```python
# To look at the first 10 rows of Data we can use
df.iloc[0: 10]
```

- This is a pretty good way of getting Data back but there are drawbacks
- If we want to get the first Column back, we would have to use the Index Notation of this as follows
    - This isn't too bad... but not great, what happens if we want the `job_work_from_home` Column?
    - We would have to Count inwards on how many Columns that is
        - Once we find it's the 6th one in, and remembering it's a Zero based Index so it'll be in position 5 then we can get the Results
- Now what happens if we want to Include Multiple Columns at the same time specifically looking at the Salary Values (`salary_rate`, `salary_year_avg`, `salary_hour_avg`)
    - This would be a mess to try and count over to them and to try and figure out where they live
- This is where the `.loc` Method comes into play and is waaaaaaaaay better!

```python
# This will Return, the first Column for the First 10 Rows
df.iloc[0:10, 0]
```

```python
# This will Return, the sixth Column for the First 10 Rows
df.iloc[0:10, 5]
```

```python
# Showcasing the .loc Method to get the Values from the Multiple Salary Columns
# Same Annotation as the iloc Method: First Part specifies the Row Index, Second Part defines the Columns
# Difference here is, we can use the Actual Column Names, instead of having to use the Index of those Columns and counting to find those Indexes. Can just reference the Columns directly
# As well as this, we can define multiple Columns and use the Range Notation (Slicing Method) to get all of the Columns back
# Get from the first one defined to the last one defined, as well as all the other ones in between
df.loc[0:10, 'salary_rate':'salary_hour_avg']
```

- Problem is we got back a whole list of NaN Values which is very Useless
- We are unable to Interpret this... so we will actually need to get Data to Analyze
    - First, Filter it to include more Values. We will expand our Search Range, instead of only grabbing 10 Rows we will grab all
        - This can be done by just having a Colon in the Index position `df.loc(:)` (This is a Selct All)
    - Next, Filter out/Drop all the NaN Values so we only get back Rows with Meaningful Data
        - IMPORTANT!: We only want to do this for a specific Column and that's `salary_rate`
        - Because, you will never have data in both `salary_hour_avg` and `salary_year_avg`. One will always be NaN, in that case, we will never get any Results if we Drop all Nan Values
        - Since typically, if there's an Hourly Rate there won't be a Yearly Rate and vice versa

```python
# Dropping all NaN Values and Running this without specifying a Column
df.loc[:, 'salary_rate':'salary_hour_avg'].dropna()
```

```python
# Dropping all NaN Values for a specific Column
df.loc[:, 'salary_rate':'salary_hour_avg'].dropna(subset='salary_rate')
```
