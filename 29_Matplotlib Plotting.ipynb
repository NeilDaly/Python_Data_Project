# Matplotlib: Plotting
___
- EG: `plt.bar(df)`
- Now it's time to learn how to use this matplotlib Library of Plotting by answering a Data Analytical Question!
- We are going to do Multiple:
    1. Line Chart: For this we will look at what is the Trend of Job Postings over time?
    2. Bar Chart: For this we will look at how many jobs are associated with each of the Job Titles?

```python
# Imports
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

# Load Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

- We are going to start with a simple Example first
    - We have 2 Variables X and Y (Which we will use appropriatly for each Axis) and they are both Lists of Numbers (1 through 4) - This will be a straight Line
    - We want to Plot this
- And ofc... there's a Method for this!!
    - `matplotlib.pyplot.plot`
    - It plots Y versus X as lines and/or Markers (This is great for either Line Charts or Scatter Plots)

```python
x = [1, 2, 3, 4]
y = [1, 2, 3, 4]

plt.plot(x, y)
# To Remove the Object Memory Location Notation, include the Following:
plt.show()
```

- Now it's time to start Printing some actual Data but we will start simple with the job_posting_date
- Let's Plot the Date Values from this entire Data Frame and we are going to Plot it against each other... so we are going to end up with a Straight Line

```python
# Showing the first 5 job_posted_dates. We get both the Date and the Time back 
df.job_posted_date.head()
```

```python
# Define the Plot Function. And then we define the X and Y and we can use the Data Frame itself!
# Because we are plotting the same Column against itself, it's going to look repetitive (since it's on both axis))
plt.plot(df.job_posted_date, df.job_posted_date)
plt.show()
```
