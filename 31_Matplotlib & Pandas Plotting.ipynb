# Matplotlib: Pandas Plotting 
___
- EG: `df.plot(kind='bar')`
- There's a slightly easier way to Plot using Pandas but we had to understand Matplotlib first before looking into the Pandas way of doing things (a shortcut)
- Let's look at the difference and how to do this via Pandas!

```python
# Imports
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Load Dataset
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
# Looking at the Job Title Postings Bar Chart from the Previous Lesson
title_counts = df.job_title_short.value_counts()
title_counts
```

```python
# Making the Job Title Postings Bar Chart from the Previous Lesson
plt.bar(title_counts.index, title_counts)
plt.title('Postings by Job Title')
plt.ylabel('Count of Job Postings')
plt.xticks(rotation=45, ha='right')
plt.show()
```

- Now ignoring all the Formatting that has been applied to the Bar Chart
    - Let's focus on the Core Code itself which is `plt.bar(title_counts.index, title_counts)`
    - Here we are Calling the Bar Function where we need to specify the Index and also the Job Counts
        - This is pretty annoying to have to remember all of this stuff (accessing the Index Value, with the correct Property)
- Luckily for us, we won't have to remember all of this!
    - Inside the Pandas Documentation for both **Series** and **DataFrames**, there is a `plot` Function
        - `pandas.Series.plot`
        - `pandas.DataFrame.plot`
    - Since we've been working with both of these Object Types, it's pertinent to mention both (They both behave similarly and have the same Properties Available to them)
        - Can Provide multiple different Argumets to the `.plot` Function since it has an Unpack Operator
        - The Parameters we can Provide to it are:
            - data: Series or DataFrame
            - X: Label or Position to be used - If it's a DataFrame, Provide the String Value of the Column
            - Y: Label or Position to be used - If it's a DataFrame, Provide the String Value of the Column
        - Also included is the kind Argument!
        - No longer do you need to memorize all the different Functions for the different Plots
        - Now you can just Type it as a String Value!

```python
# Plotting with Pandas
# Because the Object is a Series, we don't need to Provide an X or Y Value, we just need to tell it, what Graph we want
title_counts.plot(kind='bar')
```

```python
# Plotting with MatPlotLib
plt.bar(title_counts.index, title_counts)
```

- Comparing the 2 Graphs, the Pandas one looks better, is much less Code and is a whole lot simpler to implement and read
- Since the X Axis Titles aren't overlapping and works fresh out the box without having to manipulate it further
After Comparison the Pandas one is better and a whole lot simpler
- And if we want to change the Graph to a Line, we can just Update the String Value to Line and Run it and that's all that's needed!
- The cool thing is, all the Formatting that was worked on before can also be brought forward too
    - Since the Plot Module of Matplotlib is being used in the Background, we can specify all the different Labelling that we want to have, as before!

```python
title_counts.plot(kind='bar')
plt.ylabel('Count of Postings')
plt.title('Job Postings by Job Title')
plt.xticks(rotation=45, ha='right')
plt.show()
```

- That was for our `title_counts` Variable which is of Type: **Series**!
- How would we do this on a **DataFrame**?
    - Now in this case with the `.plot` Function, we will need to provide the X and Y Parameters
        - Because the DataFrame has multiple different Columns in it compared to the Series, which only had 2
        - So it needs to know what to use
- Specifically for this, let's Plot over time using a Line Chart, how the Salary is Trending over the Year in 2023

```python
# Plotting with a DataFrame: Providing the X and Y Axis Values along with the kind of Chart to Display
df.plot(x='job_posted_date', y='salary_year_avg', kind='line')
```

- After Running this, we can see that the provided Line Chart does not provide a whole lot of Value...
- We can see that the Values are Fluctuating around 200K, based on the Y Axis of Salaries, showing the Dates from the Beginning of the Year to the End of the Year
- This Example was to mainly demonstrate that we can do Plotting with DataFrames
    - The DataFrame in our case has over 700_000 Values so the Line Chart ends up showing a lot of Noise in there
    - To make this better, we'd have to take a similar approach of breaking this down on a Monthly Basis and then Plotting it
- The more important Concept is how to use **Series** and **DataFrames**
