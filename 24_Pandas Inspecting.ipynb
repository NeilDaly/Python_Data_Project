- Now, what happens if we want to access the 100th Row of this Data?
- For this we can use a **Property** of **Data Frame** (Which is somewhat similar to a **Method**)
    - This is called `pandas.DataFrame.iloc`, stands for integer location
    - We can use Integers to callout things like the Index Number of a Row and also, that of the Column (Which we will look at a bit later)

```python
# Get the 90th Record
df.iloc[90]
```

```python
# This Method also supports Slicing so we can INput a Range and get that Data back like so:
df.iloc[90:100]
```

```python
# Also Supported, a second Argument can be provided. An Integer Location can be provided of even the Column. Here we are accessing the index, job_title_short and job_title at Locations 0:2
df.iloc[90:100, 0:2]
```

- It's very rare that we will access the Column this way through Integers
- Many reasons which include it not being clear which Columns you are getting, readability, messy and physically having to count the Columns to include what you want can be confusing
- So, instead it's good to use `.iloc` to get your Row Range (90:100 in this case), and then from there, specifying the Column Names from the Data Frame that you want to Display
    - As you can see below:

```python
df[['job_title_short', 'job_location']].iloc[90:100]
```

- What if we want to inspect the Data Frame but specifically for a certain job title that we identified above?
- Specifically, we want to look at the Data Analyst Jobs
- We can see from the Data these are in Random Locations...
    - To find these we can use a comparison operator to Filter for this as seen below:
    - `df.job_title_short == 'Data Analyst'` - When Running just this we get True/False Values back so we can tell which Indexes contain the Condition we are searching for
    - Now we can use this to Filter a Data Frame
- Now all we do is Wrap it in `[]` to make it a List and name it `df` to Filter for the following Condition
    - Now it becomes a Condition that we are Evaluating, and we are putting this inside of `df`

```python
df[df.job_title_short == 'Data Analyst']
```

- Running this, we now get back all the different Data Analyst Job Postings in here
- And at the bottom, it even tells us how many Records we have Returned (196,075 Rows)

- Let's continue on with this Analysis and we don't want to just look at Data Analyst Jobs. We want to add more Filters
- We also want to look at where the Salary for these Jobs is > 100_000 because right now, we are getting back a lot of NaN Values and those aren't very Valuable
    - So let's look at how we can do this:
        - First off if we just add in the `and` Condition we will run into a bunch of Errors
        - We will need to wrap both of our Conditions in `()` to make sure both of these are Evaluated as `TRUE`
        - Running it like this, we still get Errors... and this is because of the `and` Operator
        - The `and` Operator is incorrect and we shouldn't be using this in here... instead we will need to use the `&` Operator
        - Using this, we now get our Results Returned and no longer get an Error

```python
df[(df.job_title_short == 'Data Analyst') & (df.salary_year_avg > 100_000)]
```

```python
# We can even do an OR Operator in here by using the pipe symbol '|'
# This will Display all Results that show if any of the Conditions are True, showing all jobs that are either Data Analyst (even ones without a Salary), OR if the Salary > 100_000
df[(df.job_title_short == 'Data Analyst') | (df.salary_year_avg > 100_000)]
```

```python
# What happens if we only want to Return back all the Records that contain Numerical Values for the salary_year_avg?
# We learned about the .nan() Method from the Numpy Lesson. This Returns all Values that aren't Blank. It looks at Values that are Not A Number(Blanks) and Skips them
df[(df.job_title_short == 'Data Analyst') & (df.salary_year_avg.notna())]
```

```python
# Most of the Values did have NA in there and very few actually had a Salary Attached to them
# This shows that any Record that doesn't have a Number associated (or in this case shows as NAN) is False
    # Any Record that does have a Value associated with it, would show as True
# Because we only want the Records that do have Salaries associated with them, we would want the True Values of the below Condition
df.salary_year_avg.notna()
```
