# Exercise: Cleaning Data
___
- Like any Data SOurce, the Data that we are using for the Project Requires some cleanup
- We will be using Modules from the Python Standard Library in order to clean the Data up
- In the Data we are using, most will be okay but there are 2 Columns that will cause Issue and they are: job_skills and job_posted_date
    - This is because the job_posted_date is stored as a String... but this should be a DateTime Object. Pyhon has a DateTime Object inside of it
    - Similarly, the job_skills are also stored as a String, but it should be a List
- It's because all this Data is loaded into a CSV File, looking at the job_skills, it's wrapped in Quotes, which casts it as a String
- In this Exercise, we are going to Clean up both the job_skills Column (to Convert it into a List) and the job_date Column (to Convert it into a DateTime Object)

```python
# List of Dictionaries
data_science_jobs = [
    {'job_title': 'Data Scientist', 'job_skills': "['Python', 'SQL', 'Machine Learning']", 'job_date': '2023-05-12'},
    {'job_title': 'Machine Learning Engineer', 'job_skills': "['Python', 'TensorFlow', 'Deep Learning']", 'job_date': '2023-05-15'},
    {'job_title': 'Data Analyst', 'job_skills': "['SQL', 'R', 'Tableau']", 'job_date': '2023-05-10'},
    {'job_title': 'Business Intelligence Developer', 'job_skills': "['SQL', 'PowerBI', 'Data Warehousing']", 'job_date': '2023-05-08'},
    {'job_title': 'Data Engineer', 'job_skills': "['Python', 'Spark', 'Hadoop']", 'job_date': '2023-05-18'},
    {'job_title': 'AI Specialist', 'job_skills': "['Python', 'PyTorch', 'AI Ethics']", 'job_date': '2023-05-20'}
]
```

- For this Cleanup, we will need to use the `datetime` Import: As this will convert our job_date into DateTime

```python
# Import datetime from datetime
from datetime import datetime
```

```python
# Using the Function now, we can get the current DateTime Value
datetime.now()
```

```python
# So what do we want to do with this Date?
# We want to Convert it to the correct Object it needs to be
# Let's inspect what is actually coming out by accessing one of these elements inside of our List of Dictionaries

# Because it's a List and it's Indexed, we will start by looking at the Element in Position 0
data_science_jobs[0]
```

```python
# Now we want to Access the job_date of this Element
data_science_jobs[0]['job_date']
```

```python
# Now we Verify the Type of this - We have Confirmed it's a String - We want to get it into a DateTime 
type(data_science_jobs[0]['job_date'])
```

```python
# Storing the job_date Value inside a Variable so we can Call it
test_date = data_science_jobs[0]['job_date']
```

```python
# Now we invoke the String Parse Time Method inside the DateTime Import to Convert it
datetime.strptime(test_date,'%Y-%m-%d')
```

```python
# Wrapping it in a Print Function, Prints it out in the Proper Format that we want
print(datetime.strptime(test_date,'%Y-%m-%d'))
```

```python
# What happens now if we want to go through and Clean-up all of the DateTimes?
# Because, currently, we have only done it for one... how are we going to do this?
# We will go through it using a FOR Loop and Convert the Dates from String into proper DateTime Object
# After that it will replace, the Original String Value in the Dictionary with the Converted DateTime Value
# As a Note, to Convert it back into a String, use strftime (just change the p to an f in the below)

for job in data_science_jobs:
    job['job_date'] = datetime.strptime(job['job_date'], '%Y-%m-%d')
    print(job['job_date'])
```

```python
# Printing out the List of Dictionaries, we can see that the job_date has been Converted and Replaced Successfully with the DateTime Objects
data_science_jobs
```

- Now that Task 1 is Complete (We have cleaned up those DateTime Objects)
- Now we need to work on Task 2: Clean up the Skills and Convert them into a List
- We need to find something to Convert this... looking into the Python Standard Library: we will find a Module called `ast - Abstract Syntax Trees`
- This is not a commonly used Module... or a common problem/issue we will run into or need to solve, just for demonstration purposes
- Probably won't usually need to Convert Strings into Lists, although the DateTime one is very common so we should commit that one to Memory
    - The Method we care about in the `ast` is `literal_eval`

```python
# Import the Module
from ast import literal_eval
```

```python
# Pull the job_skills from the first Dictionary entry and Assign it to a Variable
test_skill = data_science_jobs[0]['job_skills']
```

```python
# Convert it from a String into a List, using the ast Method

literal_eval(test_skill)
```

```python
# Using a FOR Loop to do it for all the job_skills in the List of Dictionaries and replace the Values currently in there with the Converted Values

for job_skills in data_science_jobs:
    job_skills['job_skills'] = literal_eval(job_skills['job_skills'])
```

```python
# Printing out the List of Dictionaries to see the Converted Values

data_science_jobs
```

- Note: You can have both Convert Conditions in the same FOR Loop
- I broke it out because I wanted to do it step by step, learn for Practice as well and do it separately to see what was happening
- To have it all together, see below:

```python
from datetime import datetime
from ast import literal_eval

for job in data_science_jobs:
    job['job_date'] = datetime.strptime(job['job_date'], '%Y-%m-%d')
    job['job_skills'] = literal_eval(job['job_skills'])

data_science_jobs
```

- Now both Tasks have been Completed!
    - Task 1: Convert the Date String into a DateTime Object
    - Task 2: Convert the job_skills String into a List Object
