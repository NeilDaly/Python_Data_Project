# List Comprehension
___
- EG: `[x for x in rnage(10)]`
- As we move forward in these Lessons, we are going to start combining a lot of the things we are learning
- In this Lesson, we will be combining Lists and Loops
- So what is this List Comprehension?

```python
# Example of List Comprehension
# In this, the List Comprehension or what's going on is inside of the square Brackets here - Which is a List
# What we are doing is generating numbers from 0 to 100

numbers = [x for x in range(100)]

# Printing out this Variable, we will see: A List of Values from 0 to 100
print(numbers)
```

```python
# So, let's have a think about what we'd have to do if we didn't know about this List Comprehension and we just needed to Generate it with the knowledge we have already...
# First thing we'd have to do is define a Blank List. This is done like so:
numbers = []

# Then from here, we'd need to Iterate through a Loop and put it into the Blank List to make it not Blank - Fill it with Values

for number in range(100):
    numbers.append(number)

print(numbers)
```

- This is the exact same thing as we have done above
- Both of these Implementations do the exact same thing and Print the exact same Results
- But when we can Accomplish in 1 Line of Code, without it would require multiple lines of Code and a lot more thinking/knowledge
- As well as this, the reasdability of both of these is very different
- The first one is much nicer, cleaner and way easier to read compared to the second one...

So now, let's go over the Syntax of this

```python
# Walking through this Step by Step:
# We will start by Defining the List Comprehension that we will actually be doing by using the [] - as this Defines a List
# The first Argument we need is the Variable that we want to put into this List so we are just going to define it as 'x'
# After the Variable, now we are going to use a For Loop to Iterate through what we want to (In our case, we wanted to Generate the numbers from 0 to 99)

numbers = [x for x in range(100)]
numbers
```

```python
# Now we will Modify this to show what's going on...
# We want to do an Operation to each one of those numbers as we are iterating through 100
# Let's say we want to Multiply them times 2
# So everytime that we cycle through each one of these Numbers, it multiplies the Value times 2, adding it to the List and then we have our Final List

numbers = [x*2 for x in range(100)]
numbers
```

```python
# We aren't limited to Math Operations either, we can do something like Converting X to a Float
# Now we have all these different Items Converted to a Float

numbers = [float(x) for x in range(100)]
numbers
```

```python
# Moving onto the Iterator itself of that Range(100)
# We aren't Limited to just using Range(100)
# We can use any Item that is Iterable... We talked about Lists, Set, Tuples, Dictionaries and things like this Range() Function are Iterable
# And even things like Strings!

letters = [x for x in 'PYTHON']
letters
```

### List Comprehension

Not Limited to only `list` Comprehension:
- `set` Comprehension
- `tuple` Comprehension
- `dictionary` Comprehension

All you have to do is actually change what's going on with the Brackets themselves

```python
# Example of doing this with a Set
# It Returns it Printed out of Order and makes it un-readable... This is because Sets are Un-Ordered
# But! It does have the Letters of python in there

letters = {x for x in 'PYTHON'}
letters
```

```python
# What's the Purpose of this List Comprehension? What is a Real World example?
# Well, think back to the last lesson when we Analysed the Massive List of Jobs (700k) to see if they contained the term 'Data Analyst' (160k) and put that back into a List
# We can basically do the same thing with List Comprehension - And the code is much cleaner, shorter and easier to Read

analyst_list = []

for job in job_list:
    if "Data Analyst" in job:
        analyst_list.append(job)
```

```python
# Define the List of analyst_list
# Starting by openeing the []
# Define a Variable so we will name the Variable job
# We are going to cycle through the job_list by saying: for job in job_list
# job_list is the name of the Column in the Data set that we are Importing
analyst_list = [job for job in job_list]
```

```python
# Now we already have defined that we can do the for in here - Now let's incorporate this IF Statement in there
# We are going to Iterate through each Item in the List, from there it's going to check if it's TRUE, if we meet this condition for this Item and if it is, put it into the List
# The below Statement does the exact same as the one 2 Cells up. We just did it in 1 Line as opposed to 4 Lines. Much cleaner and easier to read
analyst_list = [job for job in job_list if 'Data Analyst' in job]
```
