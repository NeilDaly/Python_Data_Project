# Sets
___

- EG: `{'python', 'sql', 'excel'}`
- Now time to learn about the next container Data Type called Sets
- As per Luke, it's only downhill from here...
- Sets are very similar to Lists with the Exception of, instead of using `[]` we will be using `{}` but there's no Colon `:` so it doesn't become a Dictionary
- It's as if, Lists and Dictionaries were combined to make Sets but with some small Syntax differences
- As we will see, Sets are Mutable so we can Add/Remove Items from them
- However, Sets are Unique! In the fact that if we were to convert a List into a Set, it'll remove Items that are Duplicated

| Feature          | List                              | Dictionary                             | Set            | Tuple                                    |
|------------------|-----------------------------------|----------------------------------------|----------------|------------------------------------------|
| Usage            | Collection of ordered items       | Key-value pairs                        | Unique items   | Fixed data                               |
| Syntax           | `[item1, item2, ...]`             | `{'key1': value1, 'key2': value2}`     | `{item1, ...}` | `(item1, item2, ...)` or `item,`         |
| Type of Data     | Sequence                          | Mapping                                | Set            | Sequence                                 |
| Order            | Ordered                           | Unordered                              | Unordered      | Ordered                                  |
| Indexing         | Yes (by index)                    | Yes (by key)                           | No             | Yes (by index)                           |
| Duplicate Values | Allowed                           | Values allowed, keys must be unique    | Not allowed    | Allowed                                  |
| Mutability       | Mutable                           | Mutable                                | Mutable        | Immutable                                |

- Sets are mainly designed for **unique Items**
- They don't allow **Duplicate Values**, however, they are **Unordered** and there's **No Indexing Available** with them

```python
# Let's get into defining our first Set
# We assign it a Variable Name
# Then we define it as a Set by putting in {}
# From here we put in a List of Skills
# We are only using Strings, but once again we can put any Data Types inside of a Set

job_skills = {'tableau', 'sql', 'python', 'statistics'}
# Printing them out - Notice something about it?
# When we Printed out a List, they stayed in an Ordered manner - They Printed out in the Order they were defined
# Sets don't stay Ordered... they don't Print in the Order they are defined
# Usually they will be put in Alphabetical Order but don't always count on it... don't think you can use Sets to Alphabatize things. It doesn't work like that
job_skills
```

```python
# For Lists, whenever we wanted to find an Item in a Certain position, you could look up the Index
# You would, call the Variable, define it with a square bracket, [], and include the Index to look it up!
# With Sets you can't do that, it'll throw an Error because the Set Object is not Indexed

job_skills[1]
```

```python
# Looking at what we can do with them...

help(set)
```

```python
# Alot of the same Common Methods that we used in Lists are available with Sets
# So if we wanted to Add a New Skill

job_skills.add('looker')
job_skills
```

```python
# Now let's Test the feature of only having Unique Values!
# Let's add something to this that's already in this Set
# It Runs without Problems and won't return an Error... buuut it won't be Added because it's already in there
# Looking at the Print out, we can see only 1 Instance of SQL

job_skills.add('sql')
job_skills
```

```python
# Now, let's Remove an Element from this Set
# Using the same Pop Method here, that we used in Lists
# Using the format that's below will return an Error - Reason being is because the Pop Method here takes no Arguments, but we supplied one
# Reason being is because with Lists they are Indexed and we could Provide it an Indexed Value to Remove something in it
# But with Sets because they are Unordered, they don't have Indexes

job_skills.pop('tableau')
job_skills
```

```python
# So Running it without an Argument will Remove an Arbitrary Value from the Set
# In this case it Removed: tableau
# For Luke, it Removed: statistics
# So, it can't be trusted and we have no idea what it will Remove. Bit of a Russian Roulette

job_skills.pop()
```

```python
# Printing the Set out, we can see what is left

job_skills
```

```python
# So if we wanted to Remove a specific Value from the Set... we would have to use the Remove Method
# The Remove Method will Remove the first occurrance of a Value from a Set
# This works, because there are no Dupes allowed in Sets so because we are only removing the First Instance of the Value, it will be the ONLY Instance of the Value
# So the Remove Method is safe to be used here because no Dupes are allowed and you are defining exactly what needs to be Removed from the Set
```

```python
job_skills.remove('statistics')
job_skills
```

```python
# So then, what are we using Sets for and what are their Purposes?
# The primary Use Case is being able to extract out unique Values weather it's in something like a list or a Tuple
# And being able to do that very efficiently
# Let's look at an Example
# We have this List of Skills and inside of it we have Duplicate Values

skill_list = ['python', 'sql', 'statistics', 'tableau', 'python', 'sql', 'statistics', 'tableau']

# What we can do now is convert the above List Object to a Set Object
# As we can see from the Printout, it Removed the Duplicate Values from the List efficiently and very quickly

set(skill_list)
```

```python
# Now if we want to Convert it back to a List
# We can just Wrap it again in a List Function and Running it like below - It is now back to a List

list(set(skill_list))
```
