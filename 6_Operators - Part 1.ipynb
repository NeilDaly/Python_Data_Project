```python
# Let's say we had something like the below

5 / 2
```

```python
# Right now we converted it to a Float Value - The answer is in a Float
# But we want to maintain it as an Integer
# In this case we can do something called Floor Division
# Which is using two forward slashes (//) instead of just one single forward slash (/)
# This is how we Return and Maintain an Integer Answer/Output

5 // 2
```

```python
# Alternatively if we wanted to get the Remainder of the above Division then we would use Modules
# This returns the Remainder
# So 5 / 2 - Goes in 2 times, the Remainder is 1 so that should be what is Returned

5 % 2
```

```python
# Floor Division is a pretty common Operation we would do if we'd have to maintain an Integer and not get a Float

print('5 Divided by 2 is:', 5 // 2, 'with a Remainder of:', 5 % 2)
```

### Quick Reminder on Order of Operations

1. **Parentheses:** Anything in Parentheses is executed first
2. **Exponents:** Operations unvolving Exponents (including Roots, indicated as fractional Exponents)
3. **Multiplication and Division:** Processed from left to right
4. **Addition and Subtraction:** Also Processed from left to right

```python
# The underscore is ignored, but it's a common way to separate up Values
# Makes it easy to count if it's a thousanth, millionth, billionth, etc
# More for readability and understandability to see the Value at a glance

min_salary = 20_000 # _ is ignored by the interpreter
max_salary = 80_000
avg_salary = (min_salary + max_salary) / 2

avg_salary
```

### Assignment Operators

| Operator | Name                    | Example   |
|----------|-------------------------|-----------|
| `=`      | Assign                  | `x = y`   |
| `+=`     | Add and Assign          | `x += y`  |
| `-=`     | Subtract and Assign     | `x -= y`  |
| `*=`     | Multiply and Assign     | `x *= y`  |
| `/=`     | Divide and Assign       | `x /= y`  |
| `%=`     | Modulus and Assign      | `x %= y`  |
| `**=`    | Exponent and Assign     | `x **= y` |
| `//=`    | Floor Divide and Assign | `x //= y` |

- These are called Assignment Operators
- We have already gone over the Equals sign before `=` but not the others
- These are going to combine the Arithmetic Operations along with an Assignment Operator
- All the different Arithmetic Operations are still Available
- After going through Examples of the Assignment Operators we now know how to use them
- All of the above are available to use but the most frequesntly used is the one we went over and that is **Add and Assign**
- So that is all we are going to cover in the lesson, but the others are used in the exact same way

```python
# Diving into the core Assignment Operator of Equals
# Let's say we have 2 Applicants as shown below
# Let's say we want to switch those Applicants

applicant_1 = 'Luke'
applicant_2 = 'Neil'
```

```python
# We Create a Temporary Variable and Assign it to Applicant_1

temp_applicant = applicant_1
```

```python
# Then we Assign Applicant_1 to be Equal to Applicant_2
# Then we Assign Applicant_2 to be Equal to the Temporary Applicant

applicant_1 = applicant_2
applicant_2 = temp_applicant
```

```python
# Then we now go and Print them out

applicant_1, applicant_2
```

```python
# The cool thing about Python is that you can do Assignments of Multiple Values on the same line
# This new way of doing it will of course render everything above completely useless
# The below Assignement within one line allows us to use a comma and swap Values all in the same line
# After Running the below we have now Swapped the Applicant positions
# This is pretty cool because you can't do this in other programming languages

applicant_1, applicant_2 = applicant_2, applicant_1
```

```python
applicant_1, applicant_2
```

```python
# Okay moving onto the next Assignment Operator
# Let's say we have something like the below
# From here, now we want to Add 1 to this Variable of x

x = 1
```

```python
# So we could write it out like the below
# All we are saying is now X = 1 + 1

x = x + 1
```

```python
# Printing out X we get 2

x
```

```python
# And of course, there is actually an easier way to do this and we are about to go through this
# Instead of writing out the whole formula again(X + 1), we can use the Add and Assign Operator as seen below and then put the Value in that we want to Increment by
# This is the exact same as shown two Cells up but it stops us having to write everything out again, making our code clear and easy to read

x = 1

x += 1 # This line is the same as x = x + 1 (which is what we showed up above)
```

```python
# Now when Printing x, it's = 2
# Since we are adding a value of 1 to the Value that lives in x which is 1 and making x equal the new value so we are making x equal 1 + 1 which is 2
x
```

```python
# Conversly if we wanted to multiply x by 2
# We can do the below:

x = 1
x *= 2

x
```

### Comparisson Operators

| Operator | Name                  | Example   |
|----------|-----------------------|-----------|
| `==`     | Equal to              | `x == y`  |
| `!=`     | Not equal to          | `x != y`  |
| `>`      | Greater than          | `x > y`   |
| `<`      | Less than             | `x < y`   |
| `>=`     | Greater than or equal | `x >= y`  |
| `<=`     | Less than or equal    | `x <= y`  |

- These are called Comparisson Operators
- These are the last thing we are going to cover in this Lesson
- These are vey basic and we have used them in SQL and in Excel
- It is just looking to see if Values are Equal to each other, not equal to each other, greater than or less than each other

```python
# So let's say we want to Compare 2 different Salary Variables (which is what the Project is all about)

salary_neil = 60_000
salary_luke = 40_000

# The below line will swap the Variables around and make the Else Statement Print to showcase this working. Uncomment it to see! 
#salary_neil, salary_luke = salary_luke, salary_neil
```

```python
# We want to check if Neil's Salary is Equal to Luke's Salary by using the double Equals sign. This will either Return True or False
# And the answer is False

salary_neil == salary_luke
```

```python
# It's not just limited to Variables, we can see if literals work too
# We can check if 1 == 1 and this will Return True because 1 is equal to 1
# Note: If doing comparisons, make sure to put double Equals because it won't work if you do one
# This means, trying to Assign 1 = 1 but that would throw an Error as seen below as it's not possible to Assign an Int as a Variable
# In the Error Message below it does point that out and asks if you meant to use the double Equals

1 = 1
```

```python
# On the importance of Comparison Opperators and showing how they would actually be used

if(salary_neil > salary_luke):
    print("Neil is paid more than Luke")
else:
    print("Luke is paid more than Neil")
```
