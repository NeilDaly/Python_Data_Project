# Classes: The Template of Objects
---
**Class:** A Template/Blueprint for Creating Objects (records).
As seen below:
- From the Help Function, we can see INT is a **Class**
- From the Type Function, the 90000 is an **Instance** of that **Class**

```python
type(90000)
```

```python
help(int)
```

```python
# Now we are going to get into building our own Class ourselves

#Variables
job_title = "Data Analyst"
job_location = "Canada"
job_salary = "90000"
```

```python
# Because this is very standard job data, we want to store it inside of a single Class Instance

class JobPost:
    def __init__(self, title, locaiton, salary):
        self.title = title
        self.location = locaiton
        self.salary = salary
```

```python
# This is an instance of the Class: JobPost
JobPost(job_title, job_location, job_salary)
```

# Attribute: The Variables of an Object
---
**Attribute:** A Field in a record, defined by it's Class.
Can Access these via dot notation. Steps are:
- Creating the Class
- Calling the Class and Assigning it to a Variable
- Calling the Variable and using a **.** after it

```python
# In this Class that we Created, we defined Attributes inside this Class

class JobPost:
    def __init__(self, title, locaiton, salary):
        # Below are the Attributes (Variables) of the Class
        self.title = title
        self.location = locaiton
        self.salary = salary
```

```python
# This is an instance of the Class: JobPost
# With this instance, we can Assign this to a Variable
job_1 = JobPost(job_title, job_location, job_salary)
```

```python
# Now we can Access the Attributes of job1, in this case - Title, Location and Salary by using dot notations (.)
# From the Instance of this Class, we have output 'Data Analyst'

job_1.title
```

```python
job_1.location
```

```python
job_1.salary
```

# Methods: The Functions of an Object
---
**Method:** A Function defined inside a Class that operates on it's Objects.

- The Functions of an Object.
- A Method is a Functions superpowers (What it can do)

```python
# In this Class that we Created, we defined Attributes inside this Class

class JobPost:
    def __init__(self, title, locaiton, salary):
        # Below are the Attributes (Variables) of the Class underneath this Method
        self.title = title
        self.location = locaiton
        self.salary = salary
```

```python
# The below is a Function because it's not inside of a Class... but we can turn this into a Method!
def display_info(title, location, salary):
    return print(f"JOB:       {title}\nLOCATION:  {location}\nSALARY:    ${salary:,.0f}")
```

```python
# In this Class that we Created, we defined Attributes inside this Class

class JobPost:
    def __init__(self, title, locaiton, salary):
        # Below are the Attributes (Variables) of the Class underneath this Method
        self.title = title
        self.location = locaiton
        self.salary = salary
    # We have taken the Function, brought it into the Class, indented it in Properly, Update the Variable Names
    def display_info(self):
        return print(f"JOB:       {self.title}\nLOCATION:  {self.location}\nSALARY:    ${self.salary:,.0f}")
```

```python
# Now we can Create a Variable job_1 which is an INSTANCE of the JobPost Class with our 3 Variables that we defined
job_1 = JobPost(job_title, job_location, job_salary)
```

```python
# We can see the Attribute of it by running the dot notation on it 

job_1.title
```

```python
# And we can also Run a Method on this as well
# This is going to be using that dot notation as well
# Specifically we can call the job_1. Now multiple things appear along with the display_info Funtion
# Because the display_info appears with a Cube Icon, this defines it as a Method
# We also need open/closed bracets for this as it's going to Run the Method

job_1.display_info()
```
