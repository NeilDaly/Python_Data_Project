### Variables aren't only numbers

## Example

* Create Variables to store the following information about a job posting:
    * Job ID (Integer)
    * Company Name (string)
    * Salary Rate (string, but consider how you might store this as a numerical type for calculations)
    * Job Work From Home (boolean)

```python
# Define Variables
job_id = 1
company_name = "Neil's Company"
salary_rate = 100000
job_wfh = True

# Calling the Variables
print(job_id)
print(company_name)
print(salary_rate)
print(job_wfh)
```

```python
# Define Variables
job_id = 101
company_name = "Data Analyst"
salary_rate = 125000

# Calling the Variables
print("Data Job Description")
print("++++++++++++++++++++")
print("Job ID:", job_id)
print("Company Name:", company_name)
print("Salary:", salary_rate)
print("WFH:", job_wfh)
```

# Objects: The Data Records
---
**Objects:** A data record with Fields; an instance of a Class.
- Everything is Python is an Object and that is a record of Data

```python
# Declaring this Variable makes both of the below Objects
salary = 100000

100000 # Object 1
salary # Object 2
print(salary) # Object 3 (Print Function counts as an Object)
print # Object 4

# If we want to inspect what type of Object something is we can use the type() Function by placing the Object in as the Argument
type(100000) # This looks to see what type of Object 100000 is - It returns Int

# Variable

# Function

# Class

# Method

# Attribute
```
