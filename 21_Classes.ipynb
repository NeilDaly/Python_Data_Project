Let's look at a very simple Version, implementing a List Data Type using a Class

```python
class NeilsList:
    def __init__(self):
        """Initialize an Empty List."""
        self._items = []

    def add(self, item):
        """Add an Item to the end of the List."""
        self._items.append(item)

    def __getitem__(self, index):
        """Retrieve an Item by Index."""
        return self._items[index]
    
    def __setitem__(self, index, value):
        """Set an Item at a Specific Index"""
        self._items[index] = value

    def __repr__(self):
        """Return a String Representation of the List."""
        return str(self._items)

    def __len__(self):
        """Return the length of the List."""
        return len(self._items)
```

```python
# Create an Instance of the List
my_list = NeilsList()
```

```python
# Running the List, we get a Blank List back
my_list
```

```python
# Inside the Class, we have an Add Method, all it's doing is an Append
# So let's add something to our Blank List
# To Access these Methods, all we do is the DOT Notation
my_list.add('Data Nerd')

my_list
```

```python
my_list.add('Finance Nerd')
my_list
```

```python
# Doing the Length Method, we get the Length of the List
my_list.__len__()
```

```python
# Or we can do it this way by using the Length Function. Either way works since they are doing the same thing
len(my_list)
```

- So that's a Class that we've Built for a Custom List
- Classes are really important whenever we need to bundle up a bunch of different Methods and make an Object behave in a certain way
- And we currently have a good use case for this! Specifically the Functions we wrote in the Module Lesson for calculate_salary and calculate_bonus
- Now what we can do, is Create our own Currency Data Type for this Base Salary and thus allow us to do these Functions when Called as a Method along with a few others...

We will eventually have a Class called BaseSalary with 4 different Methods inside of it

- Dunder Init, `__init__` is a magic Method. Anywhere that has a double underscore it means they are pre-defined Methods inside of Python that when the Compiler is Ran, Python knows this specific named Method has a special use case
- We define the first positional Argument as `self`: We are going to be defining multiple Methods inside of here and they all need to be able to work with each other and to basically, pass around whatever Data we have Initiated inside of this Object of BaseSalary and the way to do this... by defining the first positional Argument in `init` as `self`
- Now we need to Build in a Variable that can be passed into it as a Second Argument, we define base_salary and now we can Pass in an Argument such as `BaseSalary(100_000)`
Next Steps!
- Now we need to Pass in 2 more Variables that we are going to be Passing into this Class and that is: bonus_rate and the currency_symbol we want to use for this, so `$`
- Note: This `self` Keyword is not the Required name of the Parameter. We can make it anything we want and it will still work EG: We can change the `self` word to `obj` and there would be no difference
- But `self` is the most common name for it and is most commonly what it's named. See this more often than anything else

```python
class BaseSalary:
    def __init__(self, base_salary, bonus_rate=0.1, symbol='$'):
        # Now we Initialize the Variable of base_salary - That's why it's called Init, it initializes these Variables
        # We set the Attribute of base_salary, this is an Attribute not a Method, equal to the Variable that we are passing in of base_salary
        self.base_salary = base_salary
        self.bonus_rate = bonus_rate
        self.symbol = symbol
        self.total_salary = base_salary * (1 + bonus_rate)
        self.bonus = self.total_salary - self.base_salary
    
    def __repr__(self):
        return f'{self.symbol}{self.base_salary:,.0f}'

    def show_salary(self):
        return f'{self.symbol}{self.total_salary:,.0f}'

    def show_bonus(self):
        return f'{self.symbol}{self.bonus:,.0f}'
```

```python
# Now we give it a Variable so we can Call it
salary = BaseSalary(100_000)

# And now we want to see the base_salary so we go ahead and Call it:
# This is an Attribute, not a Method so that's why we don't include the Parenthesis
# The Attribute of base_salary is equal to 100_000 - Attributes are basically Variables
salary.base_salary
```

```python
# We can now Print out and see what Values are Contained inside these Attributes 
print(salary.base_salary)
print(salary.bonus_rate)
print(salary.symbol)
```
