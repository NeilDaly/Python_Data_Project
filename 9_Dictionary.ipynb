# Dictionary
___

- EG: `{'language' : 'python', 'library' : 'pandas'}`
- Now time to get into a more Advanced container Data Type called Dictionaries
- This Data Type is very similar to the physical form of a Dictionary
- In a Dictionary the Definition is stored based on the term itself
- Dictionaries in Pthon are very similar!

| Feature          | List                              | Dictionary                             | Set            | Tuple                                    |
|------------------|-----------------------------------|----------------------------------------|----------------|------------------------------------------|
| Usage            | Collection of ordered items       | Key-value pairs                        | Unique items   | Fixed data                               |
| Syntax           | `[item1, item2, ...]`             | `{'key1': value1, 'key2': value2}`     | `{item1, ...}` | `(item1, item2, ...)` or `item,`         |
| Type of Data     | Sequence                          | Mapping                                | Set            | Sequence                                 |
| Order            | Ordered                           | Unordered                              | Unordered      | Ordered                                  |
| Indexing         | Yes (by index)                    | Yes (by key)                           | No             | Yes (by index)                           |
| Duplicate Values | Allowed                           | Values allowed, keys must be unique    | Not allowed    | Allowed                                  |
| Mutability       | Mutable                           | Mutable                                | Mutable        | Immutable                                |

- In this case, the Dictionary is what's in between the curly Brackets
- In here we will be going over popular **Methods** including the one that we use to Identify the Keys of a Dictionary
- We will also use ones that are very farmiliar as we used them in Lists and they work the exact same

```python
# So let's create our first dictionary
# It's important to note, you can include a List in a Dictionary as well, it would be a List of Values

job_type_skills = {
    'database' : 'postgresql', 
    'language' : 'python', 
    'library' : 'pandas'
}
```

```python
job_type_skills
```

```python
# So what can we do with these Dictionaries?
# To find that out, we use our trusty Methof of...

help(dict)
```

```python
# The first one we will look at is probably the most common Method used which is a dunder method called __getitem__
# We can also call it by a very similar way for how we Index Lists
# We can do similarly an Index Method on Dictionaries
# We are calling the KEY of Database and getting the Values inside of it

job_type_skills['database']
```

```python
# This Returns the different Keys in the Dictionary

job_type_skills.keys()
```

```python
# This Returns the different Values in the Dictionary

job_type_skills.values()
```

```python
# If we wanted to Remove something from the Dictionary
# It's similar to what we did in Lists

job_type_skills.pop('library', 'pandas')
job_type_skills
```

```python
# We can Remove Items from Dictionaries
# But what happens if we want to Add ones in?
# This is where the Update Method comes into play
# We need to define a Dictionary, which we do by using the Curly Brackets, and then providing a KEY and a Value, just like before

job_type_skills.update({'cloud' : 'google cloud'})
job_type_skills
```

```python
# There is ofc an easier Method for Updating Values that should be used
# Like below, where we used the List Notation of [] to see what Values were inside of a KEY
# Similarly, we can use the List notation in order to define a new KEY and also a Value

job_type_skills['cloud']
```

```python
# This sets the Dictionary to Add a Key of Version_Control and Assign the Value of GIT in there
# Printing it out, we can see it's been Added to the KEY of version_control

job_type_skills['version_control'] = 'GIT'
job_type_skills
```
