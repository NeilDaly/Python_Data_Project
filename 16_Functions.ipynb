# User-Defined Functions
___
- Now it's time to shift focus and get into, Building our own User-Defined Functions in order to do something

```python
base_salary = 100_000
bonus_rate = 0.1

total_salary = base_salary * (1 + bonus_rate)
total_salary

# Imagine if we have a bunch of different Base Salaries and a bunch of different Rates as well that we need to feed into the above Formula
# Well this is a perfect example for a Function
```

```python
# Let's start defining this Function

base_salary = 100_000
bonus_rate = 0.1

# Any time we start a Function, it needs to start with the Keyword def (which stands for Define)
# From there, we now need to Name the Function (whatever we want)
# Then we need Parenthesis () and then a colon :
# Then to check that we wrote it right and to see if it will execute, we can add the Keyword Pass

def calculate_salary():
    pass
```

```python
# For Demonstration purposes we will put all of the Calculation into the Function and then Call it

def calculate_salary():
    base_salary = 100_000
    bonus_rate = 0.1

    total_salary = base_salary * (1 + bonus_rate)
```

```python
# Calling the Function here - Remeber to add () at the end, since it's a Function, that's what we have to Add so we can Call it
# And we get..... nothing! Why is this??
# It's because we didn't Return anything in the Function itself! We have to specify what we want to Return back from this

calculate_salary()
```

```python
# With a Return Value

def calculate_salary():
    base_salary = 100_000
    bonus_rate = 0.1

    total_salary = base_salary * (1 + bonus_rate)

    return total_salary
```

```python
# And now calling this Function we can see that we get.... an Output
# It's because we told it what to Return in the Function

calculate_salary()
```

```python
# Now we have a problem... this Function is useless
# If we want to change the Variables, we have to go into the Function itself and Update them in there/input new Variables in there
# Now we have to make the Function useable, we can Provide the Function Arguments, which will then Provides us Values

# So we are going to Remove the Variables from inside of the Function
# And now we are going to Define those Arguments inside of the Function Parenthesis
# We are going to Pass in - base_salary and bonus_rate
base_salary = 100_000
bonus_rate = 0.1

def calculate_salary(base_salary, bonus_rate):

    total_salary = base_salary * (1 + bonus_rate)
    return total_salary
```

```python
# And now calling this Function will Present an Error
calculate_salary()
```

```python
# This is because, we did not Provide/Pass in the Arguments, when Calling the Function
# We now have to provide the base_salary and bonus_rate Arguments
# Providing those and Running it, we get an Output
calculate_salary(base_salary, bonus_rate)
```

- Technical Note: The Variable Names that we are using are defined above
- We are Passing in the same Variable Names as Arguments and that's not even necessary
- We can actually put in Raw Values and still get it outputting the correct answer: See Below

```python
calculate_salary(150_000, .25)
```

```python
# If we wanted to, we can even change the Variable Names as well

salary_1 = 100_000
rate_1 = 0.1

def calculate_salary(base_salary, bonus_rate):

    total_salary = base_salary * (1 + bonus_rate)
    return total_salary
```

```python
# Passing in those Arguments, gives us:

calculate_salary(salary_1, rate_1)
```

```python
# Cleaning things up and making it more Professional and Organised, we should write it like this:
# Now we have our Function up above and then the Variables, and the Function Call together below
# It's not necessary where we have the Variables, but ofc the Function needs to be Defined before you actually call it

def calculate_salary(base_salary, bonus_rate):
    total_salary = base_salary * (1 + bonus_rate)
    return total_salary
```

```python
salary_1 = 100_000
rate_1 = 0.1

calculate_salary(salary_1, rate_1)
```

```python
# Building on this example further!
# What happens, if we routinely have the same bonus_rate? And we don't want to slow down our coding and have to put in that salary_rate or the bonus_rate every single time
# Well.... we can add what's called; An Optional Argument
# We want to Define that a Typical bonus_rate is 10%
# To do this: We will Define what that bonus_rate is right next to the Argument itself
# We put in the equals sign and then the Value we want to Assign to it, should it be Optional or not

# In this case, we are specifying it to be .1

def calculate_salary(base_salary, bonus_rate=.1):
    total_salary = base_salary * (1 + bonus_rate)
    return total_salary
```

```python
salary_1 = 100_000
rate_1 = 0.1

# Now, we are still calling the Function and only Providing 1 Argument as seen below
# When Running this, we get the Value
calculate_salary(salary_1)
```

```python
salary_1 = 100_000
rate_1 = 0.1

# No matter what we put in there, it will work
calculate_salary(200_000)
```

```python
# And in the case, where the Rate is different?
salary_1 = 100_000
rate_1 = 0.5

# It now takes in this rate_1 Argument whenever we provide it
# And it even Overrides that Optional Argument as well 
calculate_salary(salary_1, rate_1)
```
