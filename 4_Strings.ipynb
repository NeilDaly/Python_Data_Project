```python
# Explaining Dunder/Magic Methods
# Dunder and Magic Methods are the same thing just different names
# All of the below will print the same thing. Number 3 is better to use as it's cleaner and easier to read. Plus less mistakes/syntax errors with it
# The third one is doing the Dunder Add Method in the background

str.__add__('Data ', 'Analyst')
'Data'.__add__(' Analyst')
'Data' + ' Analyst'
```

```python
# We also have a Multiply Dunder Method
# But again for readability we don't need the Method, we can just Multiply it because Python is smart

'Data Analyst '.__mul__(10)
'Data Analyst' * 10
```

```python
# Getting the String Length

'Data Analyst'.__len__()
len('Data Analyst')
```
